```python
#9. 
def is_valid_bracket_sequence(s):
    # Dictionary to map opening and closing brackets
    bracket_map = {')': '(', '}': '{', ']': '['}
    stack = []

    for char in s:
        if char in bracket_map.values():  # If it's an opening bracket
            stack.append(char)
        elif char in bracket_map:  # If it's a closing bracket
            if not stack or stack[-1] != bracket_map[char]:  # Stack is empty or mismatch
                return "The given string is not valid"
            stack.pop()  # Valid match, remove the last opening bracket
        else:
            return "The given string is not valid"  # Invalid character

    return "The given string is valid" if not stack else "The given string is not valid"
 # Test cases
s = "()[]{}"
print(is_valid_bracket_sequence(s))  # Output: The given string is valid

s = "(){]"
print(is_valid_bracket_sequence(s))  # Output: The given string is not valid
```

1. You are given a large integer represented as an integer list of digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's. Increment the large integer by one and return the resulting ;list of digits. Implement the logic by creating a function.

        Example 1:
        Input: digits = [1,2,3]
        Output: [1,2,4]
        Explanation: The array represents the integer 123.
        Incrementing by one gives 123 + 1 = 124.
        Thus, the result should be [1,2,4].

        Example 2:
        Input: digits = [4,3,2,1]
        Output: [4,3,2,2]
        Explanation: The array represents the integer 4321.
        Incrementing by one gives 4321 + 1 = 4322.
        Thus, the result should be [4,3,2,2].

        Example 3:
        Input: digits = [9]
        Output: [1,0]
        Explanation: The array represents the integer 9.
        Incrementing by one gives 9 + 1 = 10.
        Thus, the result should be [1,0].
        ]

```python
def plusOne(digits):
    # Start from the last digit and add one
    digits[-1] += 1

    # Handle carry over from the last digit to the first digit
    for i in range(len(digits) - 1, -1, -1):
        if digits[i] == 10:
            digits[i] = 0
            if i != 0:
                digits[i - 1] += 1
            else:
                # If we are at the first digit, we need to add a new digit at the beginning
                digits.insert(0, 1)

    return digits

# Example usage
digits = [0, 1]
new_digits = plusOne(digits)
print(new_digits)  # Output should be [0, 2]

# Example usage
digits = [9, 9]
new_digits = plusOne(digits)
print(new_digits)  # Output should be [0, 2]
```

1. Use a custom defined timeit decorator to measure the execution time of any function. Use time module to compute time.
        
        Hint:
        import time
        current_time = time.time()

```python
# use a custom define timeit decorator to measure the execution time of any function.Use time module to compute time
import time

def timeit_dec(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        x = func(*args,**kwargs )
        end = time.time()
        print(f"The time it took to compute {func.__name__} is {end-start} seconds")
        return x
    return wrapper

@timeit_dec
def num_gen(n):
    print(f"Entering num_gen with n = {n}")
    for i in range(n):
        print(f"Printing {i}")
        print(f"Currently at iteration {i+1} out of {n}")
    print(f"Exiting num_gen")
    
@timeit_dec
def sum_of_list(input_list):
    time.sleep(3)
    sum = 0
    for i in input_list:
        sum = sum + i
    print(f"Sum of list is {sum}")
    return sum

num_gen(10)
input_list = [1,2,3,4,5]
print(sum_of_list(input_list))
```

```python
# creating a class rectangle. Methods area() and perimeter()
class Rectangle:
    def __init__(self, length , breath):
        self.length = length
        self.breath = breath
        self.calc_area = 0
        self.calc_peri = 0
    def area(self):
        self.calc_area = self.length * self.breath
    
    def perimeter(self):
        self.calc_peri = 2 * (self.length + self.breath)
    
rec1 = Rectangle(2,4)
rec1.area()
print(f"The area of the rectangle is {rec1.calc_area}")
rec1.perimeter()
print(f"The perimeter of the rectangle is {rec1.calc_peri}")

# water container returns max amount of water 
height = [1,8,6,2,5,4,8,3,7]
max_area = 0
class Container:
    def __init__(self, height):
        self.height = height
        self.max_area = 0
        self.area = 0
    def water(self):
        for i,vertical_line in enumerate(height):
            # print(i)
            # print(vertical_line)
            for next_line_idx,next_line in enumerate(height[i+1:]):
                # print(next_line_idx)
                # print(next_line)
                self.length = min(vertical_line,next_line) 
                self.breath = next_line_idx + 1
                self.area = self.length * self.breath
                self.max_area = max(self.area,self.max_area)
        return self.max_area
    
c = Container(height)
print(f"The max water in the container is {c.water()}")
```

1. I have provided the students.json inside data_files folder that contains students records. Using that json, compute following things:

* Calculate the percentage for each students assuming 100 is the full marks for each subjects. Also create a new json file with additional field "percentage" in json whose value will be the percentage you have computed.

* Then load the new json file and generate the statistics of:
    * How many students failed? (i.e. percentage < 40)
    * How many students achieve third division (i.e. percentage between 40 and 50)
    * How many students achieve second division (i.e. percentage between 50 and 65)
    * How many students achieve first division (i.e. percentage between 65 and 80)
    * How many students achieve distinction division (i.e. percentage 80 and above )



```python
import json

# Step 1: Calculate percentage for each student
with open("../data_files/students.json", "r") as my_file:
    json_content = json.load(my_file)
    for content in json_content:
        all_marks = content["marks"]
        total = sum(all_marks)
        percentage = (total / 400) * 100
        print(f"{content['name']} : {percentage} %")
        content["percentage"] = percentage

with open("../data_files/new_students.json", "w") as my_file:
    json.dump(json_content, my_file)

# Step 2: Calculate Division based on percentage
with open("../data_files/new_students.json", "r") as my_file:
    json_content = json.load(my_file)

    for content in json_content:
        if content["percentage"] >= 80:
            print(f"{content['name']} : Division : A")
        elif content["percentage"] >= 60:
            print(f"{content['name']} : Division : B")
        elif content["percentage"] >= 40:
            print(f"{content['name']} : Division : C")
        else:
            print(f"{content['name']} : Division : F")
```
