```python
import pandas as pd
```

```python
data = {
    "name":[],
    "roll":[],
    "address":[],
    "age":[]
    }
```

```python
for i in range(0,10):
    data["name"].append(input("Enter the name: "))
    data["roll"].append(int(input("Enter the roll: ")))
    data["address"].append(input("Enter the address: "))
    data["age"].append(int(input("Enter the age: ")))
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
a= df["age"].values
val=[]
for i in a:
    if i<=15:
        val.append("Junior")
    elif i<=25:
        val.append("Mid")
    else:
        val.append("Senior")
val
```

```python
df.insert(column="category",loc=0,value=val)
```

```python
df
```

```python
sum = 0
count =0
for i in df["age"]:
    sum += i
    count += 1
average_age = sum/count
print(f"Average age is: {average_age}")
```

```python
value1 = df["category"].value_counts()
print(value1)
```
