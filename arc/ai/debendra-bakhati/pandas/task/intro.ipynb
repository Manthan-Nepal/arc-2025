
### TASK 1:
#### a) Filter the DataFrame to only include rows where the Price is greater than $50

#### b) Filter the resulting DataFrame to only include rows where the Category is either "Electronics" or "Fashion"

#### c) Print the resulting DataFrame

```python
import pandas as pd
```

```python
data = {
    'Product': ['Tablet', 'Wheat','Jeans', 'Camera', 'Sneakers', 'Bluetooth Speaker', 'Sunglasses','rice'],
    'Price': [499.99,80.90, 39.99, 749.99, 79.99, 19.99, 59.99, 60],
    'Category': ['Electronics','Household', 'Fashion', 'Electronics', 'Fashion', 'Electronics', 'Fashion','Household']
}

df = pd.DataFrame(data)
df.head()
```

```python
df = df[df['Price']>50]
df
```

```python
df1 = df[(df['Category'] == 'Electronics') | (df['Category'] == 'Fashion')]
```

```python
df1.head()
```

### TASK 2:
##### a) Merge the two DataFrames on the CustomerID column
##### b) Print the resulting DataFrame

```python
data1 = {
    'CustomerID': [5, 6, 7, 8],
    'Name': ['Sita Rai', 'Hari Sharma', 'Rita Thapa', 'Kiran Gurung'],
    'Address': ['12 Naxal Rd', '34 New Baneshwor', '56 Jawalakhel', '78 Putalisadak']
}

data2 = {
    'CustomerID': [5, 6, 5, 7],
    'OrderDate': ['2024-03-01', '2024-03-05', '2024-03-12', '2024-03-15'],
    'TotalAmount': [250.00, 175.50, 320.75, 400.00]
}
```

```python
df1= pd.DataFrame(data1)
df1.head()
```

```python
df2= pd.DataFrame(data2)
df2.head()
```

```python
merge_df = pd.merge(df1,df2,on='CustomerID')
```

```python
merge_df.head()
```

TASK 3:

a) Group the DataFrame by the Region column

b) Calculate the total Sales for each region

c) Calculate the average Sales for each region

d) Print the resulting DataFrame

```python
data = {
    'Region': ['Central', 'Himalayan', 'Terai', 'Valley', 'Central', 'Himalayan', 'Terai', 'Valley'],
    'Sales': [180, 220, 270, 320, 370, 420, 480, 530],
    'Quarter': ['Q1', 'Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q2']
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
grouped = df.groupby('Region')['Sales'].agg(Total_sales = 'sum',Average_Sales = 'mean')
grouped.head()
```
