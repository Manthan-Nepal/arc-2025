```python
import json
from collections import defaultdict

with open("students.json", "r") as f:
    students = json.load(f)

top_students = []
course_totals = defaultdict(int)
course_counts = defaultdict(int)
course_max = defaultdict(lambda: ("", -1))
course_min = defaultdict(lambda: ("", 101))
address_distribution = defaultdict(int)

for student in students:
    total = sum(student["marks"])
    avg = total / len(student["marks"])
    top_students.append((avg, student["name"]))
    address_distribution[student["address"]] += 1
    for course, mark in zip(student["courses"], student["marks"]):
        course_totals[course] += mark
        course_counts[course] += 1
        if mark > course_max[course][1]:
            course_max[course] = (student["name"], mark)
        if mark < course_min[course][1]:
            course_min[course] = (student["name"], mark)

top_students.sort(reverse=True)
top_5_students = [{"name": name, "average": avg} for avg, name in top_students[:5]]

average_per_course = {course: course_totals[course] / course_counts[course] for course in course_totals}

highest_scorers = {course: {"name": name, "mark": mark} for course, (name, mark) in course_max.items()}
lowest_scorers = {course: {"name": name, "mark": mark} for course, (name, mark) in course_min.items()}

result = {
    "top_5_students": top_5_students,
    "average_per_course": average_per_course,
    "highest_scorers": highest_scorers,
    "lowest_scorers": lowest_scorers,
    "student_count_per_address": dict(address_distribution)
}

with open("students_percentage.json", "w") as f:
    json.dump(result, f, indent=4)
```
