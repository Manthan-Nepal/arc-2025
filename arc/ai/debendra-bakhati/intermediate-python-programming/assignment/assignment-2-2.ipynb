```python
#Given marks of students of a class: marks = [ 10, 20, 30, 40, 50, 60]. Calculate the average mark.

marks = [10,20,30,40,50,60]
sum = 0
count = 0
for i in marks:
    count += 1
    sum += i
average = sum/count
print(count)
print("The average is: ",average)
```

```python
#Write a program to pring multiplication table of a number given by user.

num1 = int(input("Enter a positive integer: "))
mul1 = 0
print(f"The multiplication table of {num1} is: ")
for i in range(1,11):
    mul1 = i * num1
    print(f"{num1}*{i} = {mul1}")
```

```python
#Given an integer x, return true if x is a palindrome, and false otherwise.

num2 = int(input("Enter a integer: "))

str1 = str(num2)

str2 = str1[::-1]

if str1 == str2:
    print(True)
else:
    print(False)
```

```python
#Write a program to count the number of vowels in a string.

str3 = input("Enter a string: ")
str3 = str3.lower()
count = 0
for i in str3:
    if i == "a" or i == "e" or i == "i" or i == "o" or i == "u":
        count += 1
print(f"The number of vowels in '{str3}' is {count}.")
```

```python
#Given a roman numeral, convert it to an integer. Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

s = "LVIII"
sum = 0
for i in s:
    if i == "M":
        sum += 1000
    if i == "D":
        sum += 500
    if i == "C":
        sum += 100
    if i == "L":
        sum += 50
    if i == "X":
        sum += 10
    if i == "V":
        sum += 5
    if i == "I":
        sum += 1
print(sum)
```

```python
#Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

s = "([)]"
flag1 = True
flag2 = True
flag3 = True
for i in s:
    if i == "(":
        flag1 = False
    elif i== "{":
        flag2 = False
    elif i == "[":
        flag3 = False
    if flag1 == False and i == ")":
        flag1 = True
    elif flag2 == False and i == "}":
        flag2 = True
    elif flag3 == False and i == "]":
        flag3 = True
    if flag1 == True and i == ")":
        flag1 = False
    elif flag2 == True and i == "}":
        flag2 = False
    elif flag3 == True and i == "]":
        flag3 = False
if flag3 == False or flag1 == False or flag2 == False:
    print(False)
else:
    print(True)
```

```python
#Write a program to calculate the frequency of each characters in a string.

s = input("Enter a string: ")
frequency = {}

for char in s:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character frequencies:")
print(frequency)

```
