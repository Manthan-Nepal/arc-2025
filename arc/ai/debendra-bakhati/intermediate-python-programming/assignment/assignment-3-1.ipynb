```python
def plusOne(digits):
    n = len(digits)
    for i in range(n - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        digits[i] = 0
    return [1] + digits
```

```python
print(plusOne([1, 9, 9]))    
print(plusOne([4, 3, 2, 1])) 
print(plusOne([9]))          
print(plusOne([9, 9]))       
```
