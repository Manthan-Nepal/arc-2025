### Develop a calculator to take any two numbers from user and perform the operations on those two numbers (operations to be performed are chosen by user).

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Choose an operation (+, -, *, /): ")

if operation == "+":
    result = num1 + num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "-":
    result = num1 - num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "*":
    result = num1 * num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "/":
    result = num1 / num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "%":
    result = num1 % num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "**":
    result = num1 ** num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "//":
    result = num1 // num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "==":
    result = num1 == num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "!=":
    result = num1 != num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "<":
    result = num1 < num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == "<=":
    result = num1 <= num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == ">":
    result = num1 > num2
    print(f"Result of {num1}{operation}{num2} is", result)
elif operation == ">=":
    result = num1 >= num2
    print(f"Result of {num1}{operation}{num2} is", result)
else:
    result = "Invalid operation"
    print(result)
```
