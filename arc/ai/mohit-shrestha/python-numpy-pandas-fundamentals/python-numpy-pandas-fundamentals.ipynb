```python
my_set = {1, 2, 3, 4}

# Removing an element using remove()
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

# Removing an element using discard()
my_set.discard(3)   # Attempting to remove a non-existing element using remove() (raises KeyError) but discard() doesn't raise any error
print(my_set)  # Output: {1, 4}
```

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1.union(set2) 
union_set_newmethod = set1 | set2

print(union_set)
print(union_set_newmethod)
```

```python
intersection_set = set1.intersection(set2)  # alternatively, intersection_set = set1 & set2
print(f"Intersection set: {intersection_set}")

difference_set = set1.difference(set2)  # alternatively, difference_set = set1 - set2
print(f"Difference set: {difference_set}")

symmetric_difference_set = set1.symmetric_difference(set2)  # alternatively, symmetric_difference_set = set1 ^ set2
print(f"Symmetric difference set: {symmetric_difference_set}")
```

```python
def roman_to_integer(s: str) -> int:
    values = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    total = 0
    for i in range(len(s)):
        # If this is not the last character and current is less than next, subtract it
        if i + 1 < len(s) and values[s[i]] < values[s[i + 1]]:
            total -= values[s[i]]
        else:
            total += values[s[i]]
    return total

# Example usage:
print("roman to integer:", roman_to_integer("MCMXCIX"))
```

```python
def isValid(s: str) -> bool:
    # Map closing brackets to their matching opening brackets
    bracket_map = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    
    stack = []
    
    for char in s:
        if char in bracket_map.values():
            stack.append(char)
        elif char in bracket_map:
            if not stack or stack.pop() != bracket_map[char]:
                return False
        else:
            return False
    
    return len(stack) == 0

print(isValid("()"))      
print(isValid("()[]{}"))   
print(isValid("(]"))  
```
