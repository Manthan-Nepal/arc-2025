## TASK 4

Create an empty data dictionary with keys: name, roll, address, age where all the keys store a list of values. Fill in 10 valid data samples in the dictionary asking from the user. Then convert the dictionary to dataframe and add a new column to the dataframe called ‘category’. If the age is less than or equal to 15 then, category will be ‘Junior’, if its between 15 to 25(included), category will be ‘Mid’ and for the rest - category will be ‘Senior’. Display the dataframe. Calculate the average age and most frequent category.

**Data**

Aayush Sharma,101,Kathmandu,14

Sneha Thapa,102,Pokhara,18

Ramesh Koirala,103,Biratnagar,22

Priya Adhikari,104,Lalitpur,26

Nitesh Bhandari,105,Butwal,24

Anjali Shrestha,106,Bhaktapur,15

Suman KC,107,Dharan,30

Bina Lama,108,Hetauda,12

Roshan Giri,109,Nepalgunj,27

Kritika Maharjan,110,Janakpur,20

```python
import pandas as pd
information = ["Aayush Sharma,101,Kathmandu,14",

"Sneha Thapa,102,Pokhara,18",

"Ramesh Koirala,103,Biratnagar,22",

"Priya Adhikari,104,Lalitpur,26",

"Nitesh Bhandari,105,Butwal,24",

"Anjali Shrestha,106,Bhaktapur,15",

"Suman KC,107,Dharan,30",

"Bina Lama,108,Hetauda,12",

"Roshan Giri,109,Nepalgunj,27",

"Kritika Maharjan,110,Janakpur,20"]

def predefined_input():
    data = []
    for i in range(10):
        name, roll, address, age = information[i].split(",")

        student = {
            "name": name,
            "roll": roll,
            "address": address,
            "age": age
        }

        data.append(student)

    return data
```

```python
def user_input():
    data = []
    while True:
        user_input = input("Enter name, roll, address, age: ")

        if user_input == "esc":
            break

        name, roll, address, age = user_input.split(",")

        student = {
            "name": name,
            "roll": roll,
            "address": address,
            "age": age
        }
        data.append(student)

    return data
```

```python
data_one = predefined_input()
dataframe_one = pd.DataFrame(data_one)
dataframe_one
```

```python
data_two = user_input()
dataframe_two = pd.DataFrame(data_two)
dataframe_two
```

```python
# Grouping
dataframe_one["age"] = dataframe_one["age"].astype(int)

def categorize(age: int):
    if age < 15:
        return 'junior'
    elif( age > 15) & (age <= 25):
        return 'mid'
    else:
        return 'senior'
    
dataframe_one["category"] = dataframe_one["age"].apply(categorize)

dataframe_one
```

```python
avg_age = dataframe_one["age"].mean()
print(f"Average age in dataframe: {avg_age}")
```

```python
most_freq = dataframe_one["category"].mode()[0]
print(f"The most frequent category is: {most_freq}")
```

```python
# Task 4:

# Find unique values, indices and count for 2D array

arr2d = np.array([
    [1, 2, 3, 4, 2],
    [5, 6, 7, 8, 6],
    [1, 2, 3, 4, 2],  
    [9, 10, 11, 12, 10],
    [5, 6, 7, 8, 6],   
    [13, 14, 15, 16, 14]
])

uniques, unique_idx = np.unique(arr2d, return_index=True) # Flattened
print(f"Unique Values: {uniques}\nIndices: {unique_idx}")
```

```python
# Using an axis
uniques, unique_idx, unique_count = np.unique(arr2d, return_counts=True, return_index=True, axis=0) 
print(f"Unique Values: {uniques}\nIndices: {unique_idx}\nCount: {unique_count}\n")

uniques, unique_idx = np.unique(arr2d, return_index=True, axis=1) 
print(f"Unique Values: {uniques}\nIndices: {unique_idx}")
```
