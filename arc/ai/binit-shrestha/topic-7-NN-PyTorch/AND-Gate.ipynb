### AND Gate Implementation Using PyTorch

```python
import torch
import torch.nn as nn

# Set up data
X = torch.tensor([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]], dtype=torch.float32)

y = torch.tensor([[0],
                  [0],
                  [0],
                  [1]], dtype=torch.float32)

# Model Class
class ANDClass(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_one = nn.Linear(2,1)

    def forward(self, x):
        return torch.sigmoid(self.layer_one(x))
```

```python
ModelV0 = ANDClass()
ModelV0.state_dict()
```

```python
# Loss function and Optimizer
lossV0 = nn.BCELoss()
optimizerV0 = torch.optim.SGD(ModelV0.parameters(), lr=0.1)

# Model Training
epochs = 1000
for epoch in range(epochs):
    ModelV0.train()

    optimizerV0.zero_grad()
    y_pred = ModelV0(X)

    loss = lossV0(y_pred, y)
    loss.backward()

    optimizerV0.step()

    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1} of {epochs}], Loss: {loss.item():.4f}')

ModelV0.eval()
with torch.no_grad():
    predicted = ModelV0(X)
    predicted_cls = (predicted > 0.5).float()
    accuracy = (predicted_cls == y).sum() / y.shape[0]
    print(f'Accuracy: {accuracy.item()*100:.2f}%')
```

```python
torch.save(ModelV0.state_dict(), 'Models/AND-Model.pth')

loaded_model = ANDClass()
loaded_model.load_state_dict(torch.load('Models/AND-Model.pth'))
loaded_model.eval()

with torch.no_grad():
    test_input = torch.tensor([[0, 1]], dtype=torch.float32)
    prediction = loaded_model(test_input)
    input_list = [int(x) for x in test_input.squeeze().tolist()]
    print(f'Prediction for input {input_list}: {1 if prediction.item() > 0.5 else 0}')
```
