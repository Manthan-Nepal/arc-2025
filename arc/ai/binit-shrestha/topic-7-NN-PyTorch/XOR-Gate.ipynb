### XOR Gate Implementation in PyTorch

```python
import torch
import torch.nn as nn

# Set up data
X = torch.tensor([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]], dtype=torch.float32)

y = torch.tensor([[0],
                  [1],
                  [1],
                  [0]], dtype=torch.float32)

# Model Class
class XORClass(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_one = nn.Linear(2,3)
        self.layer_two = nn.Linear(3,1)

    def forward(self, x):
        x = torch.sigmoid(self.layer_one(x))
        x = torch.sigmoid(self.layer_two(x))

        return x
```

```python
torch.manual_seed(50)
ModelV1 = XORClass()
ModelV1.state_dict()
```

```python
# Loss function and Optimizer
lossV1 = nn.BCELoss()
optimizerV1 = torch.optim.SGD(ModelV1.parameters(), lr=0.1)

# Model Training
epochs = 5000
for epoch in range(epochs):
    ModelV1.train()

    optimizerV1.zero_grad()
    y_pred = ModelV1(X)

    loss = lossV1(y_pred, y)
    loss.backward()

    optimizerV1.step()

    if (epoch+1) % 200 == 0:
        print(f'Epoch [{epoch+1} of {epochs}], Loss: {loss.item():.4f}')

ModelV1.eval()
with torch.no_grad():
    predicted = ModelV1(X)
    predicted_cls = (predicted > 0.5).float()
    accuracy = (predicted_cls == y).sum() / y.shape[0]
    print(f'Accuracy: {accuracy.item()*100:.2f}%')
```

```python
torch.save(ModelV1.state_dict(), 'Models/XOR-Model.pth')
loaded_model = XORClass()
loaded_model.load_state_dict(torch.load('Models/XOR-Model.pth'))
loaded_model.eval()

with torch.no_grad():
    test_input = torch.tensor([[0, 1]], dtype=torch.float32)
    prediction = loaded_model(test_input)
    input_list = [int(x) for x in test_input.squeeze().tolist()]
    print(f'Prediction for input {input_list}: {1 if prediction.item() > 0.5 else 0}')

# One neuron not enough to represent XOR, not linearly separable
```
