# 🧠 Contrastive Learning Explained

## 🔍 What is Contrastive Learning?

**Contrastive Learning** is a self-supervised learning technique that learns to **distinguish between similar and dissimilar data** points. It is based on the idea that **similar samples should have closer representations (embeddings)**, while **dissimilar ones should be far apart** in the feature space.

It is widely used in **computer vision**, **natural language processing**, and **representation learning**.

---

## 🧩 Core Concepts

### 1. **Anchor, Positive, and Negative Samples**
- **Anchor**: The reference sample.
- **Positive**: A sample similar to the anchor.
- **Negative**: A dissimilar sample.

Goal:  
$$
\text{distance}(f(\text{anchor}), f(\text{positive})) \ll \text{distance}(f(\text{anchor}), f(\text{negative}))
$$  
Where \( f(\cdot) \) is the encoder or representation function.

---

## 🛠️ How It Works

### 1. **Data Augmentation**
For self-supervised learning:
- Create two augmented views of the same image/text → **positive pair**.
- Use other instances in the batch as **negatives**.

### 2. **Embedding Generation**
Use an encoder (e.g., CNN, Transformer) to convert data into embeddings:
- \( z_i = f(x_i) \)

### 3. **Contrastive Loss**
Measures how well the model can separate positives from negatives.

---

## 🎯 Common Loss Functions

### 1. **Contrastive Loss (Siamese Networks)**
For pairs:
$$
L = y \cdot ||z_i - z_j||^2 + (1 - y) \cdot \max(0, m - ||z_i - z_j||)^2
$$  
- \( y = 1 \) if similar, 0 otherwise  
- \( m \) = margin

---

### 2. **Triplet Loss**
Uses anchor, positive, and negative:
$$
L = \max(0, ||z_a - z_p||^2 - ||z_a - z_n||^2 + \alpha)
$$  
- \( \alpha \) = margin

---

### 3. **NT-Xent Loss (SimCLR)**
Normalized temperature-scaled cross-entropy loss:
$$
L_{i,j} = -\log \frac{\exp(\text{sim}(z_i, z_j)/\tau)}{\sum_{k=1}^{2N} \mathbb{1}_{[k \ne i]} \exp(\text{sim}(z_i, z_k)/\tau)}
$$  
- \( \text{sim}(z_i, z_j) = \frac{z_i^\top z_j}{||z_i|| \cdot ||z_j||} \)
- \( \tau \) = temperature parameter

---

## 🔧 Architectures

### 1. **Siamese Networks**
- Two identical encoders with shared weights.
- Compare outputs of two inputs.

### 2. **Triplet Networks**
- Three branches: anchor, positive, and negative.

### 3. **Contrastive Frameworks**
- **SimCLR**: Uses NT-Xent loss, large batch sizes.
- **MoCo**: Momentum encoder and memory bank.
- **BYOL**: Learns without negative pairs.
- **Barlow Twins**: Uses redundancy reduction.

---

## 📦 Applications

- **Image retrieval**
- **Face recognition**
- **Text similarity**
- **Unsupervised pretraining**
- **Multi-modal learning** (e.g., CLIP)

---

## 🧪 Example: SimCLR Pipeline

1. Apply two augmentations on each image: \( x \rightarrow x_1, x_2 \)
2. Encode: \( z_1 = f(x_1), z_2 = f(x_2) \)
3. Project to latent space: \( h_1 = g(z_1), h_2 = g(z_2) \)
4. Compute contrastive loss across batch using NT-Xent

---

## ✅ Benefits

- Does not need labeled data.
- Learns powerful representations.
- Generalizes well to downstream tasks.

## ❌ Limitations

- Needs large batch sizes (e.g., SimCLR).
- Computationally intensive.
- Choosing good negative samples is tricky (solved by BYOL, etc.)
