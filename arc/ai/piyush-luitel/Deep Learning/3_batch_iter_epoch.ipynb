## 🔁 Batch, Iteration, and Epoch in Deep Learning

### 🧱 1. Batch
- A **batch** is a subset of the training data.
- Instead of passing the entire dataset at once, data is split into smaller batches to train the model.
- Common batch sizes: 8, 16, 32, 64, etc.

### 🔂 2. Iteration
- **One iteration** is when the model processes **one batch** of data **and updates the weights**.
- So, if you have 100 samples and a batch size of 20:
  - You'll have **5 iterations per epoch** (100 ÷ 20 = 5).

### 🔁 3. Epoch
- **One epoch** means the model has seen **all samples in the training dataset once**.
- During one epoch, multiple iterations occur (depending on the batch size).
- More epochs = better learning (to a point), but risk of overfitting.

---

### 📌 Example:
Assume:
- Training samples: **100**
- Batch size: **20**
- Epochs: **3**

Then:
- ✅ **Each epoch** = 5 iterations (100 ÷ 20)
- ✅ **Total iterations** = 5 × 3 = 15
