# Important Activation Functions in Deep Learning

## 1. Sigmoid

- **Formula:**  
  $$ \sigma(x) = \frac{1}{1 + e^{-x}} $$
- **Range:** (0, 1)
- **Description:**  
  Sigmoid squashes input values into the range between 0 and 1, making it suitable for modeling probabilities (especially in the output layer of binary classifiers).
- **Pros:**  
  - Smooth gradient which helps in gradient-based optimization.  
  - Outputs values interpreted as probabilities.
- **Cons:**  
  - Prone to vanishing gradients for very large positive or negative inputs, slowing down learning.  
  - Output is not zero-centered, which can slow convergence.

---

## 2. Hyperbolic Tangent (tanh)

- **Formula:**  
  $$ \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$
- **Range:** (-1, 1)
- **Description:**  
  The tanh function maps inputs to values between -1 and 1, centering the data around zero, which often leads to faster convergence during training compared to sigmoid.
- **Pros:**  
  - Zero-centered output helps optimization by balancing positive and negative inputs.  
  - Steeper gradient than sigmoid, thus often preferred in hidden layers.
- **Cons:**  
  - Still suffers from vanishing gradients for very large or very small inputs.

---

## 3. ReLU (Rectified Linear Unit)

- **Formula:**  
  $$ \mathrm{ReLU}(x) = \max(0, x) $$
- **Range:** [0, ∞)
- **Description:**  
  ReLU outputs zero for negative inputs and passes positive inputs unchanged. It is computationally efficient and helps mitigate the vanishing gradient problem, making it the default choice for many deep networks.
- **Pros:**  
  - Sparse activation (outputs zero for half of the inputs), leading to efficient computation and reduced overfitting.  
  - Helps with faster convergence due to non-saturating gradient.
- **Cons:**  
  - "Dying ReLU" problem: neurons can get stuck outputting zero and stop learning if inputs are always negative.

---

## 4. Leaky ReLU

- **Formula:**  
  $$
  \mathrm{LeakyReLU}(x) = 
  \begin{cases}
    x & \text{if } x > 0 \\
    \alpha x & \text{if } x \leq 0
  \end{cases}
  $$
  where \(\alpha\) is a small positive constant (e.g., 0.01)
- **Range:** (-∞, ∞)
- **Description:**  
  Leaky ReLU modifies ReLU by allowing a small, non-zero gradient for negative inputs, which helps keep neurons active and mitigates the dying ReLU problem.
- **Pros:**  
  - Allows learning for negative inputs, reducing dead neurons.  
  - Retains benefits of ReLU like computational efficiency.
- **Cons:**  
  - Introduces a small slope in the negative region, which might not always be optimal.  
  - Requires tuning the leakage parameter \(\alpha\).

---

## 5. Softmax

- **Formula:**  
  $$ \mathrm{Softmax}(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}} $$
- **Range:** (0, 1), with outputs summing to 1 across all classes.
- **Description:**  
  Softmax converts a vector of raw scores (logits) into probabilities that sum to 1, making it ideal for multi-class classification in the output layer.
- **Pros:**  
  - Provides a probabilistic interpretation of output scores.  
  - Useful for multi-class problems where each input belongs to exactly one class.
- **Cons:**  
  - Sensitive to outliers in logits (large logits dominate the output).  
  - Not used in hidden layers; only suitable for output layers.

## Summary Table

| Activation | Range       | Use Case                 | Key Advantages                   | Main Drawbacks                  |
|------------|-------------|--------------------------|---------------------------------|--------------------------------|
| Sigmoid    | (0, 1)      | Binary classification    | Probabilistic output, smooth    | Vanishing gradients, not zero-centered |
| Tanh       | (-1, 1)     | Hidden layers            | Zero-centered, steeper gradient | Vanishing gradients             |
| ReLU       | [0, ∞)      | Hidden layers            | Efficient, sparse activations   | Dying ReLU problem             |
| Leaky ReLU | (-∞, ∞)     | Hidden layers            | Fixes dying ReLU, efficient     | Requires parameter tuning      |
| Softmax    | (0, 1), sum=1 | Output layer (multi-class) | Probabilistic multi-class output | Sensitive to logits, not for hidden layers |
