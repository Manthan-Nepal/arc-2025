```markdown
# Comparison of CNN Architectures: LeNet, AlexNet, VGGNet, ResNet, and InceptionNet

## 1. LeNet (LeNet-5)

**Developed By**: Yann LeCun et al. in 1998  
**Purpose**: Digit recognition (MNIST dataset)

### Architecture
- **Input**: 32x32 grayscale image
- **Layers**:
  1. Convolutional Layer (C1): 6 filters of size 5x5, stride 1
  2. Subsampling Layer (S2): Average pooling, 2x2 filter, stride 2
  3. Convolutional Layer (C3): 16 filters of size 5x5
  4. Subsampling Layer (S4): Average pooling
  5. Fully Connected Layer (C5)
  6. Fully Connected Layer (F6)
  7. Output Layer: Softmax classifier

### Strengths
- Pioneered CNNs, demonstrating local receptive fields and parameter sharing.
- Efficient for small datasets and simple tasks.

### Limitations
- Not scalable to high-resolution images.
- Uses average pooling (less effective than max pooling in modern architectures).
- Shallow and lacks generalization ability for complex tasks.

---

## 2. AlexNet

**Developed By**: Alex Krizhevsky et al. in 2012  
**Purpose**: Image classification (ImageNet competition)

### Architecture
- **Input**: 227x227x3 RGB image
- **Layers**:
  1. Conv1: 96 filters, 11x11 kernel, stride 4
  2. Max Pooling: 3x3, stride 2
  3. Conv2: 256 filters, 5x5 kernel
  4. Max Pooling
  5. Conv3: 384 filters, 3x3 kernel
  6. Conv4: 384 filters, 3x3 kernel
  7. Conv5: 256 filters, 3x3 kernel
  8. Max Pooling
  9. FC6: 4096 neurons
  10. FC7: 4096 neurons
  11. Output: 1000-way softmax

### Strengths
- Introduced ReLU activation for faster convergence.
- Used Dropout to reduce overfitting.
- Utilized GPU computing for training large-scale models.

### Limitations
- High computational cost.
- Manually designed filters and architecture.
- Large number of parameters (~60 million).

---

## 3. VGGNet (VGG-16 / VGG-19)

**Developed By**: Simonyan and Zisserman in 2014

### Architecture
- **Input**: 224x224x3 RGB image
- **Layers**:
  - Convolutional blocks of 3x3 filters, increasing depth:
    - VGG-16: 13 convolutional layers + 3 FC layers
    - VGG-19: 16 convolutional layers + 3 FC layers
  - Max Pooling after every block
  - FC6: 4096 neurons
  - FC7: 4096 neurons
  - FC8: 1000-way softmax

### Strengths
- Very uniform architecture: only 3x3 conv layers and 2x2 pooling
- Depth improves performance over AlexNet
- Transfer learning: widely used in pre-trained models

### Limitations
- Very high number of parameters (~138 million)
- Slow to train
- Requires a lot of memory and computational power

---

## 4. ResNet (Residual Network)

**Developed By**: He et al. in 2015

### Architecture
- **Input**: 224x224x3 RGB image
- **Key Concept**: Residual connections or "skip connections"
- **Typical Models**: ResNet-18, ResNet-34, ResNet-50, ResNet-101
- **Building Block**: Residual block = F(x) + x, where F(x) is a stack of conv layers

### Strengths
- Solves the vanishing gradient problem
- Enables training of very deep networks (over 100 layers)
- State-of-the-art performance in many tasks

### Limitations
- More complex design compared to VGG/AlexNet
- Training still computationally intensive

---

## 5. InceptionNet (GoogLeNet / Inception-v1, v2, v3, v4, Inception-ResNet)

**Developed By**: Szegedy et al. at Google in 2014

### Architecture
- **Input**: 224x224x3 RGB image (varies by version)
- **Key Concept**: Inception modules that use multiple kernel sizes (1x1, 3x3, 5x5) in parallel
- **Inception-v1 (GoogLeNet)**:
  - 22 layers deep
  - Inception modules with 1x1 convolutions for dimensionality reduction
  - Global average pooling instead of FC layers

### Strengths
- Computationally efficient despite depth
- Good balance of depth and performance
- Parameter efficiency

### Limitations
- Complex architecture with many hyperparameters
- Inception modules are not intuitive to design manually

---

## Summary Table

| Architecture   | Year | Depth | Parameters | Key Innovation             | Strengths                            | Limitations                              |
|----------------|------|-------|------------|----------------------------|----------------------------------------|------------------------------------------|
| LeNet          | 1998 | 7     | ~60K       | Local receptive fields     | Simple, fast for small images         | Not suitable for large, complex images   |
| AlexNet        | 2012 | 8     | ~60M       | ReLU, Dropout              | GPU usage, better performance         | Large model, overfitting risk            |
| VGGNet         | 2014 | 16/19 | ~138M      | 3x3 conv layers            | Simple, widely adopted, effective     | High memory usage, slow training         |
| ResNet         | 2015 | 18-152| ~25-60M+    | Residual connections       | Deep learning breakthrough            | Design complexity                        |
| InceptionNet   | 2014+| 22+   | ~5-25M     | Inception modules          | Efficient, high performance           | Hard to manually tweak                   |
```
