```python
import pandas as pd

data = {
'Product': ['Laptop', 'Shirt', 'Smartphone', 'Shoes', 'Headphones', 'Watch'],
'Price': [999.99, 29.99, 599.99, 89.99, 199.99, 49.99],
'Category': ['Electronics', 'Fashion', 'Electronics', 'Fashion', 'Electronics', 'Fashion']
}

df=pd.DataFrame(data)
# Filter the DataFrame to only include rows where the Price is greater than $50
df = df[df['Price'] > 50]
# Filter the resulting DataFrame to only include rows where the Category is either "Electronics" or "Fashion"
df= df[df['Category'].isin(['Electronics', 'Fashion'])]
# Print the resulting DataFrame
print(df)
```

```python
import pandas as pd
data1 = {
    'CustomerID': [1, 2, 3, 4],
    'Name': ['John Doe', 'Jane Smith', 'Alice Johnson', 'Bob Brown'],
    'Address': ['123 Elm St', '456 Oak St', '789 Pine St', '101 Maple St']
}
df1 = pd.DataFrame(data1)
df1
```

```python

data2 = {
    'CustomerID': [1, 2, 1, 3],
    'OrderDate': ['2024-01-01', '2024-01-05', '2024-01-10', '2024-01-15'],
    'TotalAmount': [100.50, 200.75, 150.25, 300.00]
}
df2 = pd.DataFrame(data2)
df2
```

```python
pd.merge(df1, df2, on='CustomerID')
```

TASK 3:
Group the DataFrame by the Region column
Calculate the total Sales for each region
Calculate the average Sales for each region
Print the resulting DataFrame
Create a sample DataFrame
data = {
'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
'Sales': [150, 200, 250, 300, 350, 400, 450, 500],
'Quarter': ['Q1', 'Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q2']
}

```python
import pandas as pd
# Step 1: Create the sample DataFrame
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Sales': [150, 200, 250, 300, 350, 400, 450, 500],
    'Quarter': ['Q1', 'Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q2']
}
df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Region')
```

```python
summary=grouped['Sales'].agg(Total_Sales='sum',Average_Sales='mean',Max_Sales='max',Min_Sales='min').reset_index()
print(summary)
```
