```python
# 3. Given an integer x, return true if x is a palindrome, and false otherwise.
def is_palindrome(x):
    str_x = str(x)
    return str_x == str_x[::-1]

x = is_palindrome(121)

if x== True:
    print("palindrome")
else:
    print("not palindrome")
```

```python
# 5. Given a list of integers x, find the peak elements.
x = [1, 2, 3, 1, 10, 8]
peaks = []
for i in range(len(x)):
    if (i == 0 and x[i] > x[i + 1]) or \
       (i == len(x) - 1 and x[i] > x[i - 1]) or \
       (0 < i < len(x) - 1 and x[i] > x[i - 1] and x[i] > x[i + 1]):
        peaks.append(x[i])
print("Peaks:", peaks)
```

```python
# 7. Write a program to find the sum of digits in an integer.
def sum_of_digits(n):
    return sum(int(digit) for digit in str(n))

sum_digits = sum_of_digits(12345)
print("Sum of digits:", sum_digits)
```

```python
# 8. Given a roman numeral, convert it to an integer.
def roman_to_int(s):
    roman_map = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    total = 0
    prev = 0
    for char in reversed(s):
        val = roman_map[char]
        if val < prev:
            total -= val
        else:
            total += val
        prev = val
    return total

s = "LVIII"
print("Integer value:", roman_to_int(s))
```

```python
# 9. Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.
def is_valid(s):
    stack = []
    pair = {')': '(', ']': '[', '}': '{'}
    for char in s:
        if char in pair.values():
            stack.append(char)
        elif char in pair:
            if not stack or stack[-1] != pair[char]:
                return False
            stack.pop()
    return not stack

s = input("Enter bracket string: ")
print("Is valid:", is_valid(s))
```

```python
# 10. Write a program to calculate the frequency of each character in a string.
s = "hello"
frequency = {}
for char in s:
    frequency[char] = frequency.get(char, 0) + 1
print("Frequency:", frequency)
```

```python
# 11. Write a program to generate percentage of each student.
grades_dict = {
    "Alice": {"Math": 90, "Science": 85, "Literature": 88},
    "Bob": {"Math": 78, "Science": 82, "Literature": 80},
    "Charlie": {"Math": 92, "Science": 91, "Literature": 94}
}

for student, subjects in grades_dict.items():
    total = sum(subjects.values())
    percentage = total / len(subjects)
    print(f"{student}: {percentage:.2f}")
```
