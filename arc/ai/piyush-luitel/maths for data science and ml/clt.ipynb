```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
np.random.seed(42)

# Step 1: Generate a very skewed population (Exponential Distribution)
population = np.random.exponential(scale=2.0, size=100000)

# Step 2: Plot the original population distribution
plt.figure(figsize=(12, 4))
plt.subplot(1, 1, 1)
sns.histplot(population, bins=50, kde=True, color='orange')
plt.title("Original Population Distribution (Highly Skewed)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Step 3: Central Limit Theorem - show sample means for various sample sizes
sample_sizes = [5, 30, 100]
num_samples = 1000  # Number of samples to draw for each n

# Step 4: Plot sample mean distributions for each sample size
plt.figure(figsize=(18, 5))

for i, n in enumerate(sample_sizes, 1):
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=True)
        sample_means.append(np.mean(sample))

    plt.subplot(1, 3, i)
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue')
    plt.title(f"Distribution of Sample Means (n = {n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

plt.suptitle("Central Limit Theorem Demonstration", fontsize=16)
plt.tight_layout()
plt.show()

```
