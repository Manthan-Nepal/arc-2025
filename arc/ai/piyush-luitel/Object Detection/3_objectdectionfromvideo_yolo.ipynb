```python
import cv2
import os

video_path = 'WhatsApp Video 2025-06-08 at 17.56.03_5fdadb7f.mp4'
output_folder = 'frames'
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = int(fps // 3)  # 3 frames per second

frame_num = 0
saved_num = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    if frame_num % frame_interval == 0:
        cv2.imwrite(f'{output_folder}/frame_{saved_num:03d}.jpg', frame)
        saved_num += 1
    frame_num += 1

cap.release()
```
