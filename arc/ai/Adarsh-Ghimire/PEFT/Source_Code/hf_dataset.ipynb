```python
from datasets import Dataset
from datasets import load_dataset
```

```python
def parse_and_format_math_cot_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # Split the content by the [INST] tag to separate individual problem-solution pairs
    raw_entries = content.strip().split('[INST]')

    formatted_conversations = []

    for entry_text in raw_entries:
        if not entry_text.strip():
            continue # Skip empty strings

        parts = entry_text.split('[/INST]', 1) # Split into problem part and solution part

        if len(parts) == 2:
            problem_question = parts[0].strip() # This is the "Question: ..." part
            solution_cot = parts[1].strip()     # This is "Thought:...Answer:..."

            conversation = [
                {"from": "user", "value": f"{problem_question}"},
                {"from": "assistant", "value": solution_cot}
            ]
            
            formatted_conversations.append({
                "conversations": conversation,
                "source": "custom_math_cot",
                "category": "math",
                "category_score": 1.0,
                "topic": "grade5_math_cot",
                "topic_score": 1.0
            })
        else:
            print(f"Skipping malformed entry: {entry_text[:100]}...")

    return formatted_conversations


file_path = 'Data/math_cot_dataset.txt' 
processed_data_for_hf = parse_and_format_math_cot_data(file_path)
```
