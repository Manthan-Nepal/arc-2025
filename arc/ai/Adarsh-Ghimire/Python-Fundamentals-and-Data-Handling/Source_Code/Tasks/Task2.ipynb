```python
import pandas as pd
```

```python
data = {
    "name": [],
    "roll": [],
    "address": [],
    "age": []
}
```

```python
## taking input from user
for i in range(10):
    print(f"Enter detail for student {i + 1}:")
    name = input("Enter name: ")
    roll = input("Enter roll number: ")
    address = input("Enter address: ")
    age = input("Enter age: ")
    print("\n")
    print(f"Entered name for student {i + 1} is {name}")
    print(f"Entered roll number for student {i + 1} is {roll}")
    print(f"Entered address for student {i + 1} is {address}")
    print(f"Entered age for student {i + 1} is {age}")
    print("\n")
    data["name"].append(name)
    data["roll"].append(roll)
    data["address"].append(address)
    data["age"].append(age)
```

```python
df = pd.DataFrame(data)
```

```python
df_cleaned = df.drop_duplicates().reset_index(drop=True)
df_cleaned
```

```python
## adding category column based on age

def categorize_age(age):
    if age <= 15:
        return 'Junior' 
    elif age <= 25:
        return 'Mid'
    else:
        return 'Senior'
```

```python
df_cleaned['age'] = df_cleaned['age'].astype(int)
```

```python
df_cleaned['category'] = df_cleaned['age'].apply(categorize_age)
```

```python
df_cleaned
```

```python
## calculating average age
average_age = df_cleaned['age'].mean()
print(f"The average age of the students is: {average_age:.2f}")
```

```python
## most frequent category
most_frequent_category = df_cleaned['category'].mode()[0]
print(f"The most frequent category is: {most_frequent_category}")
```
