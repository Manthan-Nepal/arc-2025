```python
## outlier removal using IQR

def remove_outliers(df,column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

filtered_df = df.copy()
columns =  ['Sales','Profit','Discount','Quantity']
for col in columns:
    filtered_df = remove_outliers(filtered_df,col)

print("Shape after outlier removal:", filtered_df.shape)
```

```python

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) 
ax = ax.flatten() 

for i, col in enumerate(columns):
    sns.boxplot(x=filtered_df[col], ax=ax[i])
    ax[i].set_title(f"Boxplot of {col} after IQR removal")
    ax[i].grid(True)

plt.tight_layout()
plt.show()

```

```python
## outlier capping
def cap_outliers(df, column):
    lower = df[column].quantile(0.01)
    upper = df[column].quantile(0.99)
    df[column] = np.where(df[column] < lower, lower, df[column])
    df[column] = np.where(df[column] > upper, upper, df[column])
    return df

filtered_df = df.copy()
for col in columns:
    filtered_df = cap_outliers(filtered_df, col)
```

```python

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) 
ax = ax.flatten() 

for i, col in enumerate(columns):
    sns.boxplot(x=filtered_df[col], ax=ax[i])
    ax[i].set_title(f"Boxplot of {col} after Outlier Capping")
    ax[i].grid(True)

plt.tight_layout()
plt.show()

```

```python
df['Sales_log'] = np.log1p(df['Sales'])
df['Profit_log'] = np.log1p(df['Profit'])

# Visualize
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Sales_log'])
plt.title("Boxplot of Sales after Log Transformation")
plt.grid(True)
plt.show()
```

```python
## median imputation for outliers
def impute_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median = df[column].median()
    df[column] = np.where((df[column] < lower) | (df[column] > upper), median, df[column])
    return df

filtered_df = df.copy()
for col in columns:
    filtered_df = impute_outliers(filtered_df, col)
```

```python

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) 
ax = ax.flatten() 

for i, col in enumerate(columns):
    sns.boxplot(x=filtered_df[col], ax=ax[i])
    ax[i].set_title(f"Boxplot of {col} after Median Imputations of Outliers")
    ax[i].grid(True)

plt.tight_layout()
plt.show()

```

```python
## Z-score method
def z_score_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    z_scores = ((df[column] - mean) / std).abs()
    return df[z_scores < threshold]

filtered_df = df.copy()
for col in columns:
    filtered_df = z_score_outliers(filtered_df, col)

print("Shape after Z-score filtering:", filtered_df.shape)
```

```python

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) 
ax = ax.flatten() 

for i, col in enumerate(columns):
    sns.boxplot(x=filtered_df[col], ax=ax[i])
    ax[i].set_title(f"Boxplot of {col} after Z-score Filtering of Outliers")
    ax[i].grid(True)

plt.tight_layout()
plt.show()

```
