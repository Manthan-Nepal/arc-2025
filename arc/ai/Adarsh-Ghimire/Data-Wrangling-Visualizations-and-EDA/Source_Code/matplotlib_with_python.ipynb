## Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([0,1,2,3,4,5,6])
```

```python
## lets plot using value x and y
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
## using object-oriented approach
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
```

```python
type(fig), type(ax)
```

```python
fig,ax = plt.subplots(figsize=(6,6))
ax.plot(x,y)
ax.set(title="Sample plt",xlabel='Sample x value',ylabel='Sample y value')
fig.savefig("./sample_plot.png")
```

```python
import numpy as np

x = np.linspace(0,20,50)
x[25:]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,x**2)
```

```python
fig,ax = plt.subplots(3,1,figsize=(10,10))
ax[0].scatter(x,np.sin(x))
ax[1].plot(x,x**2)
ax[2].scatter(x,np.exp(x))
fig.suptitle("Overall Title")
ax[0].set_title("Subplot 1 Title")
ax[1].set_title("Subplot 2 Title")
ax[2].set_title("Subplot 3 Title")
```

```python
## bar plots
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title='Nut Butter Store',xlabel='Butter',ylabel='Price')
```

```python
fig,ax = plt.subplots()
ax.barh(nut_butter_prices.keys(),nut_butter_prices.values())
```

```python
rng = np.random.default_rng(seed=42)
x = rng.random(1000)

fig,ax = plt.subplots()
ax.hist(x)
```

```python
x = np.random.randn(1000)

fig,ax = plt.subplots()
ax.hist(x)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0][0].plot(x,x/2)
ax[0][1].scatter(np.random.random(10),np.random.random(10))
ax[1][0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1][1].hist(np.random.randn(1000))
```

##### Plotting data directly with pandas

```python
import pandas as pd
```

```python
data = {
    'Name': ['Anil', 'Bina', 'Chetan', 'Deepa', 'Eshan', 'Fatima', 'Gopal', 'Hema'],
    'Age': [25, 28, 35, 30, 45, 40, 29, 32],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'HR', 'Finance', 'Marketing', 'Marketing'],
    'Salary': [40000, 55000, 65000, 62000, 70000, 58000, 52000, 49000],
    'Experience': [2, 4, 10, 8, 15, 12, 5, 6],
    'Performance': [7.8, 8.6, 9.2, 8.0, 9.0, 8.9, 7.5, 7.9],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range('2025-05-27',periods=1000))
```

```python
ts
```

```python
ts.cumsum()
```

```python
ts.cumsum().plot(kind='line')
```

```python
dept_salary = df.groupby('Department')['Salary'].mean()
dept_salary.plot(kind='bar')
plt.title('Average Salary by Department')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(df['Experience'],df['Salary'])
plt.title('Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
```

```python
plt.hist(df['Age'], bins=5,edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```
