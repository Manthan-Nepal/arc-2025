```python
categorical_var = new_df.dtypes[new_df.dtypes == 'object'].index
categorical_var
```

```python
## now finding the number of unique values for each categorical values
for category in categorical_var:
    print(f"Number of unique values in {category} = {len(new_df[category].unique())}")
```

Here we can see that various columns like Order ID, Order Date, Ship Date, Customer ID,Customer Name,City,State,Product ID ,Sub-Category, Product Name have large number of distinct values , one-hot encoding will not be efficient. So for that we will drop these columns and along with it here we see that Country column consists of only USA so we will drop it too.

```python
new_df_cleaned = new_df.drop(["Order ID", "Order Date", "Ship Date", "Customer ID", "Customer Name", "City", "State", "Product ID", "Sub-Category", "Product Name","Country"], axis=1)
```

```python
new_df_cleaned.head()
```

```python
new_df_cleaned.shape
```

```python
categorical_var = new_df_cleaned.dtypes[new_df_cleaned.dtypes == 'object'].index
categorical_var
```

```python
new_df_cleaned_encoded = pd.get_dummies(new_df_cleaned,columns=categorical_var,drop_first=True)
new_df_cleaned_encoded
```

```python
## label encoding 
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
new_df_copy = new_df_cleaned.copy()

new_df_copy['Ship Mode'] = label_encoder.fit_transform(new_df_copy['Ship Mode'])
new_df_copy.head()
```

```python
## Saving the cleaned dataset
new_df_cleaned_encoded.to_csv("./Superstore_cleaned.csv",index=False)
```
