```python
import torch
import torch.nn as nn
import torch.optim as optim
```

creating training data


```python
X = torch.tensor([[0., 0.],
                  [0., 1.],
                  [1., 0.],
                  [1., 1.]])
y = torch.tensor([[0.], [0.], [0.], [1.]])
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.linear(x))
```

```python
model = SimpleNN()
```

 nn.Linear(2,1) basically performs a linear combination of the two inputs:<br>
z = w1x1 + w2x2 + b


nn.Sigmoid() applies a nonlinear activation to squash the output between 0 and 1:

This is exactly the formulation of binary logistic regression — a linear classifier trained using a sigmoid activation and binary cross-entropy loss.

```python
# Loss function and optimizer
criterion = nn.BCELoss()  # Binary Cross Entropy for binary output
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

# 📌 What is BCE Loss?

**Binary Cross-Entropy (BCE) Loss** measures the difference between the **true labels** and the **predicted probabilities** in binary classification problems.

It is used when:
- The output is a probability between 0 and 1 (using `Sigmoid`)
- The target labels are either 0 or 1

---

## 🔢 BCE Loss Function Formula

Given:

- $y \in \{0, 1\}$ : true label  
- $\hat{y} \in (0, 1)$ : predicted probability

The **BCE loss** is defined as:

$$
\text{BCE}(y, \hat{y}) = - \left( y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y}) \right)
$$

---

### ⚙️ Intuition Behind the Formula:

- If $y = 1$, then:

  $$
  \text{Loss} = -\log(\hat{y})
  $$

  → We want $\hat{y}$ to be **close to 1** for low loss

- If $y = 0$, then:

  $$
  \text{Loss} = -\log(1 - \hat{y})
  $$

  → We want $\hat{y}$ to be **close to 0** for low loss

---





```python
# Training loop
for epoch in range(1000):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()
```

```python
# Accuracy calculation
with torch.no_grad():
    predictions = model(X)
    predicted = (predictions > 0.5).float()
    accuracy = (predicted == y).sum().item() / y.size(0)
    print("AND Gate Accuracy:", accuracy * 100, "%")
    print("Loss:", loss.item())
```

```python
# XOR Data
X_xor = torch.tensor([[0., 0.],
                      [0., 1.],
                      [1., 0.],
                      [1., 1.]])
y_xor = torch.tensor([[0.], [1.], [1.], [0.]])
```

```python
xor_model = SimpleNN()
xor_model
```

```python
criterion = nn.BCELoss()
optimizer = optim.SGD(xor_model.parameters(), lr=0.1)
```

```python
# Training loop
for epoch in range(10000):
    optimizer.zero_grad()
    outputs = xor_model(X_xor)
    loss = criterion(outputs, y_xor)
    loss.backward()
    optimizer.step()
```

```python
with torch.no_grad():
    predictions = xor_model(X_xor)
    predicted = (predictions > 0.5).float()
    accuracy = (predicted == y_xor).sum().item() / y_xor.size(0)
    print("XOR Gate Accuracy:", accuracy * 100, "%")
    print("Loss:", loss.item())
```
