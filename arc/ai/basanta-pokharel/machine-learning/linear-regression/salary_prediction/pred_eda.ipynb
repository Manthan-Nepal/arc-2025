# Univariate Analysis

we only have numerical data types

```python
bins = [i for i in range(0, 12)]  

plt.figure(figsize=(8, 5))
plt.hist(df['YearsExperience'], bins=bins, edgecolor='black', color='skyblue')
plt.title('Distribution of Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.xticks(bins) 
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

1) More no of people have about 3-5 years of experience
2) Very few people(1) have 7 - 6 yrs of experience

```python
sns.boxplot(df['YearsExperience'])
```

1) No outliers in the data

2) The median line appears to be closer to the bottom (Q1) than the top (Q3).The upper whisker is longer than the lower whisker.

3) Data might be slightly right-skewed (positively skewed).



checking skewness

```python
df['YearsExperience'].skew()
```

Positive value : Indicates right-skewness.

Value is small (< 0.5) → Means the skew is mild and the data is approximately symmetric.

For salary

```python
bins = list(range(0, 130000, 10000))  


plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=bins, color='blue', edgecolor='black')

plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.xticks(bins, rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

1) most of the people have around 50 - 70 k salary


```python
sns.boxplot(df['Salary'])
```

1) No outliers detected

2) slight Pos skewness


```python
df['Salary'].skew()
```

Positive value : Indicates right-skewness.

Value is small (< 0.5) → Means the skew is mild and the data is approximately symmetric.
