You are given a large integer represented as an integer list of digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's. Increment the large integer by one and return the resulting ;list of digits. Implement the logic by creating a function.

 Example 1:
 Input: digits = [1,2,3]
 Output: [1,2,4]
 Explanation: The array represents the integer 123.
 Incrementing by one gives 123 + 1 = 124.
 Thus, the result should be [1,2,4].

 Example 2:
 Input: digits = [4,3,2,1]
 Output: [4,3,2,2]
 Explanation: The array represents the integer 4321.
 Incrementing by one gives 4321 + 1 = 4322.
 Thus, the result should be [4,3,2,2].

 Example 3:
 Input: digits = [9]
 Output: [1,0]
 Explanation: The array represents the integer 9.
 Incrementing by one gives 9 + 1 = 10.
 Thus, the result should be [1,0].

```python
def incrementing_list_of_digits(a):
    n = len(a) -1
    for i in range(n,-1,-1):
        if a[i] != 9:
            a[i] +=1
            return a
        else:
            a[i] = 0
    
    return [1] + a

print(incrementing_list_of_digits([1, 2, 3]))   # [1, 2, 4]
print(incrementing_list_of_digits([4, 3, 2, 1]))  # [4, 3, 2, 2]
print(incrementing_list_of_digits([9]))         # [1, 0]
print(incrementing_list_of_digits([9, 9]))      # [1, 0, 0]
   
```

Use a custom defined timeit decorator to measure the execution time of any function. Use time module to compute time.

 Hint:

 current_time = time.time()

```python
import time
def timeit(func):
    def wrapper(*args,**kwargs):
        current_time = time.time()
        result = func(*args,**kwargs)
        end_time = time.time()
        time_taken = end_time-current_time
        print(f'Execution time: {time_taken:.10f} seconds')
        return result
    return wrapper

@timeit
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
print(is_prime(10_000_019))
```
