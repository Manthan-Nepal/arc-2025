## Assignment 2.1<br>
Develop a calculator to take any two numbers from user and perform the operations on those two numbers (operations to be performed are chosen by user).

Add as much functionality as you could that we have learnt till now.

```python
#lets first add all the functional components 
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    if b==0:
        return "Error: cant divide by zero "
    return a/b
def modulus(a,b):
    return a%b
def power(a,b):
    return a**b
    
```

```python
#not lets implement the calculator 
def calculator():
    print("Welcome to the Simple Calculator: ")
    while True:
        print("sum : 1")
        print("Difference : 2")
        print("Multiply : 3")
        print("Divide : 4")
        print("Modulus : 5")
        print("Power : 6")
        print("Exit: 7")

        choice = int(input("please chose the operation"))

        if choice == 7:
            print("Exiting-------------")
            return 
        if choice in (1,2,3,4,5,6):
            print("Enter two numbers:")
            num1 = float(input("Enter first operand"))
            num2 = float(input("Enter second operand"))
            if choice == 1:
                result = add(num1,num2)
            if choice == 2:
                result = subtract(num1,num2)
            if choice == 3:
                result = multiply(num1,num2)
            if choice == 4:
                result = divide(num1,num2)
            if choice == 5:
                result = modulus(num1,num2)
            if choice == 6:
                result = power(num1,num2)
            
            print("Result : ",result)
            print("---------------------------------------------------------")
        else:
            print("Invalid input:")
            print("---------------------------------------------------------")
```

```python
#3 
def check_palindrome(a):
    temp = a
    rev_num = 0
    while temp > 0:
        end_digit = temp % 10
        rev_num = rev_num * 10 + end_digit
        temp = temp // 10
    if a==rev_num:
        print("The number is palindrome")
    else:
        print("The number is not palindrome")

x = int(input("Enter any number:"))
check_palindrome(x)
```

```python
#4 
list_num = [3,2,4,3,6,3,9,7,14,2,56]
# max(list_num)
def find_greatest(num):
    res = num[0]
    for temp in num:
        if temp > res:
            res = temp
    return res

find_greatest(list_num)
```

```python
#5
def find_peak_elements(x):
    peak_elements = []
    for idx,num in enumerate(x):
        if (idx == 0):
            if num > x[1]:
                peak_elements.append(num)
        elif (idx == (len(x)-1)):
            if num > x[len(x)-2]:
                peak_elements.append(num)
        else:
            if (num >= x[idx-1] and num >= x[idx+1]):
                peak_elements.append(num)
    return peak_elements

x = [1, 2, 3, 1, 10, 8]
print("Peak elements are:",find_peak_elements(x))
            


```

```python
#8 Addition rule : big small => big + small
# Subtraction rule : small big =>  big-small (can only come in pair,cant be more)
lookup = {
    'I' : 1,
    'V' : 5,
    'X' : 10,
    'L' : 50,
    'C' : 100,
    'D' : 500,
    'M' : 1000
}

def convert_to_int(str):
    str = str.upper()
    num = 0
    i =0
    while(i < len(str)):
        if (i!= (len(str) - 1) and (lookup[str[i]] < lookup[str[i+1]])):
            num += lookup[str[i+1]] - lookup[str[i]]
            i +=2
        else:
            num += lookup[str[i]]
            i +=1
        
    return num
    
print(convert_to_int("III"))     # 3
print(convert_to_int("IV"))      # 4
print(convert_to_int("IX"))      # 9
print(convert_to_int("LVIII"))   # 58
print(convert_to_int("MCMXCIV")) # 1994
print(convert_to_int("MMMCMXCIX")) # 3999 (largest valid Roman numeral)




```

```python
#9 

def check_valid_bracket_string(str):
    mapping = {
    ')':'(',
    '}':'{',
    ']':'['
    }
    input_str = []
    for ch in str:
        if ch in ('(','{','['):
            input_str.append(ch)
        elif ch in (')','}',']'):
            if not input_str:
                return False
            if input_str[-1] != mapping[ch]:
                return False
            input_str.pop()
        else:
            return False
        
    if not input_str:
        return True
    else:
        return False
    
print(check_valid_bracket_string("()"))        # True
print(check_valid_bracket_string("()[]{}"))    # True
print(check_valid_bracket_string("(]"))        # False
print(check_valid_bracket_string("([)]"))      # False
print(check_valid_bracket_string("{[]}"))      # True

```

```python
#10 
def frequency(str):
    freq = {}
    for ch in str:
        if ch not in (freq.keys()):
            freq[ch] = 1
        else:
            freq[ch] += 1
    return freq

frequency("basanta")
```

```python
#11 
def calc_percentage(d):
    percentage = {}
    for key,value in d.items():
        sum = 0 
        for i,(_ , value_in) in enumerate(value.items()):
            sum += value_in
        percentage[key] = sum / (i+1)
    
    return percentage

grades_dict = {
        "Alice": {"Math": 90, "Science": 85, "Literature": 88},
        "Bob": {"Math": 78, "Science": 82, "Literature": 80},
        "Charlie": {"Math": 92, "Science": 91, "Literature": 94}
}
calc_percentage(grades_dict)
```
