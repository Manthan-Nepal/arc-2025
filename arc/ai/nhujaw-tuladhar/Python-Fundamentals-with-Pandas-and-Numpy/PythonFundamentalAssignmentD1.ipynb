```python
import pandas as pd

data = {
    "name": [],
    "roll": [],
    "address": [],
    "age": []
}

for i in range(3):
    print("Student", i+1)
    name = input("Name: ")
    roll = input("Roll: ")
    address = input("Address: ")
    age = int(input("Age: "))

    data["name"].append(name)
    data["roll"].append(roll)
    data["address"].append(address)
    data["age"].append(age)

df = pd.DataFrame(data)

category = []
for a in df["age"]:
    if a <= 15:
        category.append("Junior")
    elif a <= 25:
        category.append("Mid")
    else:
        category.append("Senior")

df["category"] = category

print("\nStudent Data:")
print(df)

average_age = df["age"].mean()
print("\nAverage Age:", round(average_age, 2))

most_freq_category = df["category"].mode()[0]
print("Most Frequent Category:", most_freq_category)
```
