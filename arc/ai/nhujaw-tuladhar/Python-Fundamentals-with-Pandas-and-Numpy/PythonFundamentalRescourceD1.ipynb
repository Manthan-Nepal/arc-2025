```python
#8 .Given a roman numeral, convert it to an integer. Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.
s = "LVIII"
values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

total = 0
i = 0

while i < len(s):
    if i + 1 < len(s) and values[s[i]] < values[s[i + 1]]:
        total = total + values[s[i + 1]] - values[s[i]]
        i = i + 2
    else:
        total = total + values[s[i]]
        i = i + 1

print("The integer value is:", total)
```

```python
 #9 Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.
s = input("Enter brackets: ")
stack = []
valid = True

for char in s:
    if char == '(' or char == '{' or char == '[':
        stack.append(char)
    elif char == ')':
        if not stack or stack[-1] != '(':
            valid = False
            break
        stack.pop()
    elif char == '}':
        if not stack or stack[-1] != '{':
            valid = False
            break
        stack.pop()
    elif char == ']':
        if not stack or stack[-1] != '[':
            valid = False
            break
        stack.pop()

if stack:
    valid = False

if valid:
    print("Valid")
else:
    print("Invalid valid")
```

```python
#10. Write a program to calculate the frequency of each characters in a string.
s = input("Enter a string: ")

frequency = {}

for char in s:
    if char in frequency:
        frequency[char] = frequency[char] + 1
    else:
        frequency[char] = 1

print(frequency)
```

### **Assignment 3.1**
You are given a large integer represented as an integer list of digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's. Increment the large integer by one and return the resulting ;list of digits. Implement the logic by creating a function.

 Example 1:
 Input: digits = [1,2,3]
 Output: [1,2,4]
 Explanation: The array represents the integer 123.
 Incrementing by one gives 123 + 1 = 124.
 Thus, the result should be [1,2,4].

 Example 2:
 Input: digits = [4,3,2,1]
 Output: [4,3,2,2]
 Explanation: The array represents the integer 4321.
 Incrementing by one gives 4321 + 1 = 4322.
 Thus, the result should be [4,3,2,2].

 Example 3:
 Input: digits = [9]
 Output: [1,0]
 Explanation: The array represents the integer 9.
 Incrementing by one gives 9 + 1 = 10.
 Thus, the result should be [1,0].

```python
def plus_one(digits):
    n = len(digits)

    # Start from the last digit and move backward
    for i in range(n-1, -1, -1):
        if digits[i] < 9:
            digits[i] = digits[i] + 1
            return digits
        else:
            # If digit is 9, set it to 0 and continue to next digit
            digits[i] = 0

    # If all digits were 9, we have a carryover like 999 + 1 = 1000
    digits.insert(0, 1)
    return digits

# Example usage:
print(plus_one([1, 2, 3]))    # Output: [1, 2, 4]
print(plus_one([4, 3, 2, 1])) # Output: [4, 3, 2, 2]
print(plus_one([9]))          # Output: [1, 0]
```

```python
#1. Create a class Rectangle. Define the required object attributes/state yourself. Define class methods: area() and perimeter() and use them by creating an instance of rectangle.
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)
rect = Rectangle(10, 5)

print("Area of rectangle:", rect.area())
print("Perimeter of rectangle:", rect.perimeter())
```

## **Assignment 5.1**
Using the json, compute following things:
Calculate the percentage for each students assuming 100 is the full marks for each subjects. Also create a new json file with additional field "percentage" in json whose value will be the percentage you have computed.

Then load the new json file and generate the statistics of:

How many students failed? (i.e. percentage < 40)
How many students achieve third division (i.e. percentage between 40 and 50)
How many students achieve second division (i.e. percentage between 50 and 65)
How many students achieve first division (i.e. percentage between 65 and 80)
How many students achieve distinction division (i.e. percentage 80 and above )

```python
import json
with open("/content/students.json", "r") as file:
    students = json.load(file)

for student in students:
    marks = student["marks"]
    total = 0  # Initialize inside loop
    for mark in marks:
        total += mark
    percentage = total / len(marks)
    student["percentage"] = round(percentage, 2)
with open("/content/students_with_percentage.json", "w") as file:
    json.dump(students, file, indent=4)

with open("/content/students_with_percentage.json", "r") as file:
    updated_students = json.load(file)


divisions = {
    "Failed": 0,
    "Third Division": 0,
    "Second Division": 0,
    "First Division": 0,
    "Distinction": 0
}


for student in updated_students:
    percent = student["percentage"]
    if percent < 40:
        divisions["Failed"] += 1
    elif percent < 50:
        divisions["Third Division"] += 1
    elif percent < 65:
        divisions["Second Division"] += 1
    elif percent < 80:
        divisions["First Division"] += 1
    else:
        divisions["Distinction"] += 1

print("Division Statistics:")
for division, count in divisions.items():
    print(f"{division}: {count}")
```
