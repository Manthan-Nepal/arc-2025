```python
## Task: From a, select elements that are ddivisible by 2
nums = np.array([5, 6, 7, 8, 9, 10, 11, 12])
final = []

for num in nums:
    if num % 2 == 0:
        final.append(num)

print(final)
```

```python
## Task: From a, select elements that are greater than 2 and less than or equal to 10
check=[]
for num in nums:
    if num > 2 and num < 10:
        check.append(num)
print(check)
```

```python
##Get unique elements from an array 
a = np.array([11, 11, 12, 13, 14, 15, 16, 17, 12, 13, 11, 14, 18, 19, 20])
unique_values = np.unique(a)
print(unique_values)
     
```

```python
## Get indice along with unique element


a = np.array([11, 11, 12, 13, 14, 15, 16, 17, 12, 13, 11, 14, 18, 19, 20])
unique_values, indices_list = np.unique(a, return_index=True)
print(f"Unique Values: {unique_values}")
print(f"Indices List: {indices_list}")
     
```

```python
##For 2D array getting unique
data = np.array([[1, 2, 5, 6, 7, 7 ], [1, 0, 3, 4, 5, 6], [2, 1, 0, 8, 9, 5]])
np.unique(data)
```
