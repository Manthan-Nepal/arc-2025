```python
from ultralytics import YOLO

# Load a model
model = YOLO("/content/yolo11n.pt")  # load an official model
model = YOLO("/content/runs/detect/train/weights/best.pt")  # load a custom model

# Predict with the model
results = model("/content/sample1.jpg")  # predict on an image

# Access the results
for result in results:
    xywh = result.boxes.xywh  # center-x, center-y, width, height
    xywhn = result.boxes.xywhn  # normalized
    xyxy = result.boxes.xyxy  # top-left-x, top-left-y, bottom-right-x, bottom-right-y
    xyxyn = result.boxes.xyxyn  # normalized
    names = [result.names[cls.item()] for cls in result.boxes.cls.int()]  # class name of each box
    confs = result.boxes.conf  # confidence score of each box
    import matplotlib.pyplot as plt

# Get the plotted image
im = results[0].plot()  # returns a numpy array (BGR format)

# Convert BGR to RGB
im_rgb = im[..., ::-1]

# Show it using matplotlib
plt.imshow(im_rgb)
plt.axis('off')
plt.title("YOLO Prediction")
plt.show()
```
