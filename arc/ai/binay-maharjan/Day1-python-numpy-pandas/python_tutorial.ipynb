Numeric Data Types and Operations


```python
# Addition operation
x = 10
y = 20
z = x + y
print(z)
print(type(z))
```

```python
# Modulo operation for remainder
a = 10.5
b = 2
c = a % b
print(c)
print(type(c))
```

```python
# Exponent operation
x = 10.0
y = 2
z = x ** y
print(z)
print(type(z))
```

```python
# Division operation
a, b = 31, 2
c = a/b
print(c)
print(type(c))
```

```python
# Floor Division operation
a, b = 10, 3  # a = 10, b = 3
c = a // b
print(c)
print(type(c))
```

```python
# Complex operation
x, y = 2, 3
z = complex(2, 3)
print(z)
```

```python
za=complex(5, 7) 
print(z+za) #addition in complex numbers
```

Boolean Data Type

```python
# Logical Operations on Boolean variables
a = True
b = False
result_and = a and b
result_or = a or b
result_not = not a
result_xor = a ^ b
print(result_and)
print(result_or)
print(result_not)
print(result_xor)
```
