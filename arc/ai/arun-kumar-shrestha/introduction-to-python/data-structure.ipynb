# Assignment 2.1
* Develop a calculator to take any two numbers from user and perform the operations on those two numbers (operations to be performed are chosen by user). 

* Add as much functionality as you could that we have learnt till now.

```python
x = float(input('Enter the first number: '))
y = float(input("Enter the second number: "))
operation = input("Enter the operation (+, -, *, /): ")

if operation == ('+'):
    result = x + y
elif operation == ('-'):
    result = x - y
elif operation == ('*'):
    result = x * y  
elif operation == ('/'):
    if not y:
        print('Division by zero not allowed')
        exit(0)
    
    result = x / y        
else:
    print("Enter the valid operation.")
    exit(0)

print(f"The result is: {result}")
```

```python
# Question 1: Calculate the average mark
marks = [10, 20, 30, 40, 50, 60]

# Calculate the average
average = sum(marks) / len(marks)

print(f"The average mark is: {average}")
```

```python
# Question 2: Print multiplication table of a number
n = int(input("Enter a number to display its multiplication table: "))

print(f"Multiplication table for {n}:")
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```

```python
# Question 3: Check if a number is a palindrome
def is_palindrome(x):
    x_str = str(x)
    return x_str == x_str[::-1]

# Test with some examples
test_numbers = [121, 123, 1221, 12321]
for num in test_numbers:
    print(f"{num} is a palindrome: {is_palindrome(num)}")
```

```python
# Question 4: Find the greatest number in a list
numbers = [23, 45, 12, 78, 34, 56]

max_number = max(numbers)
print(f"The greatest number in {numbers} is: {max_number}")
```

```python
# Question 5: Find peak elements in a list
def find_peaks(nums):
    peaks = []
    
    if len(nums) < 3:
        return peaks
    
    for i in range(1, len(nums) - 1):
        if nums[i] > nums[i-1] and nums[i] > nums[i+1]:
            peaks.append(nums[i])
    
    return peaks

x = [1, 2, 3, 1, 10, 8]
peaks = find_peaks(x)
print(f"Input list: {x}")
print(f"Peak elements: {peaks}")
```

```python
# Question 6: Count the number of vowels in a string
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    
    for char in s:
        if char in vowels:
            count += 1
    
    return count

test_strings = ["Hello World", "Python Programming", "AI"]
for s in test_strings:
    print(f"'{s}' has {count_vowels(s)} vowels")
```

```python
# Question 7: Find the sum of digits in an integer
def sum_of_digits(n):
    total = 0
    for digit in str(n):
        total += int(digit)
    return total

numbers = [123, 456, 789, 1234]
for num in numbers:
    print(f"Sum of digits in {num} is: {sum_of_digits(num)}")
```

```python
# Question 8: Convert Roman numeral to integer
def roman_to_int(s):
    roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    result = 0
    
    for i in range(len(s)):
        if i < len(s) - 1 and roman_dict[s[i]] < roman_dict[s[i+1]]:
            result -= roman_dict[s[i]]
        else:
            result += roman_dict[s[i]]
    
    return result

roman_numerals = ["III", "LVIII", "MCMXCIV"]
for roman in roman_numerals:
    print(f"Roman numeral {roman} = {roman_to_int(roman)}")
```

```python
# Question 9: Check if a string with brackets is valid
def is_valid_parentheses(s):
    stack = []
    brackets_map = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack.pop() != brackets_map[char]:
                return False
    
    return len(stack) == 0

test_cases = ["()", "()[]{}", "(]", "{[]}", "([)]"]
for case in test_cases:
    print(f"'{case}' is valid: {is_valid_parentheses(case)}")
```

```python
# Question 10: Calculate the frequency of each character in a string
def char_frequency(s):
    freq = {}
    
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    
    return freq

# Test with examples
test_string = "hello"
print(f"Character frequency in '{test_string}':\n{char_frequency(test_string)}")
```

```python
# Question 11: Calculate percentage for each student
grades_dict = {
    "Alice": {"Math": 90, "Science": 85, "Literature": 88},
    "Bob": {"Math": 78, "Science": 82, "Literature": 80},
    "Charlie": {"Math": 92, "Science": 91, "Literature": 94}
}

for student, subjects in grades_dict.items():
    scores = list(subjects.values())
    avg = sum(scores) / len(scores)
    print(f"{student}: {avg:.2f}")
```
