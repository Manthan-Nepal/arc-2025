```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load dataset

```python
user_ratings_df = pd.read_csv("./Dataset/ratings.csv")
user_ratings_df.head()
```

```python
user_ratings_df.drop('timestamp', axis=1, inplace=True)
```

```python
movie_metadata = pd.read_csv("./Dataset/movies.csv")
movie_metadata.head()
```

```python
movie_data = user_ratings_df.merge(movie_metadata, on='movieId')
movie_data.head()
```

## Make User-Item Matrix

```python
# Clean ratings data: remove duplicates and filter movieIds
user_ratings_df = user_ratings_df.drop_duplicates(subset=["userId", "movieId"])
user_ratings_df = user_ratings_df[user_ratings_df["movieId"].isin(movie_metadata["movieId"])]
print(f"Unique users: {user_ratings_df['userId'].nunique()}, Unique movies: {user_ratings_df['movieId'].nunique()}")
if user_ratings_df['movieId'].nunique() > 10000:
    print("Warning: Large number of unique movies. Consider using a sparse matrix for scalability.")
```

```python
user_item_matrix = ratings_data.pivot(index=['userId'], columns=['movieId'], values='rating').fillna(0)
user_item_matrix
```
