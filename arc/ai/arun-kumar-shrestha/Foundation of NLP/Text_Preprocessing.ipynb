# Text Cleaning

## Importing data

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "spam.csv"

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "uciml/sms-spam-collection-dataset",
  file_path,
  pandas_kwargs={"encoding": "ISO-8859-1"}
)

df.head()
```

```python
df.rename(columns={'v1': 'Category', 'v2': 'Message'}, inplace=True)
df.head()
```

## Converting to lowercase

```python
import pandas as pd

df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df.head()
```

## remove url using regex

```python
import re

url_pattern = re.compile(f'https?://\S+')

def remove_url(text):
    return url_pattern.sub('', text)

df['Message'] = df['Message'].apply(remove_url)
df.head()
```

## Remove no word and non whitespace character

```python
df = df.replace(to_replace=f'[^\w\s]', value='', regex=True)
df.head()
```

## Remove digits

```python
df = df.replace(to_replace=r'\d', value='', regex=True)
df.head()
```

## Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

df['Message'] = df['Message'].apply(word_tokenize)
df.head()
```

## stop word removal

```python
from nltk.corpus import stopwords 
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
df['Message'] = df['Message'].apply(lambda x: [word for word in x if word not in stop_words])
df.head()
```

## stemming/lemmatization

```python
from nltk.stem import PorterStemmer
import pandas as pd

stemmer = PorterStemmer()

def stem_words(words):
    return [stemmer.stem(word) for word in words]

df['stemmed_messages'] = df['Message'].apply(stem_words)
df.head()
```

```python
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()

def lemmatize_tokens(tokens):
    def get_wordnet_pos(word):
        tag = nltk.pos_tag([word])[0][1][0].upper()
        tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV}
        

        return tag_dict.get(tag, wordnet.NOUN)
    
    lemmas = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]

    return lemmas

df['lemmatized_messages'] = df['Message'].apply(lemmatize_tokens)
df.head()
```
