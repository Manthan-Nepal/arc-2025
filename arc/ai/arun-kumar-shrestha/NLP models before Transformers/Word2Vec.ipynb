```python
from gensim.models import Word2Vec
```

## Load dataset

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "Corona_NLP_test.csv"

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "datatattle/covid-19-nlp-text-classification",
  file_path,
)

df.head()
```

```python
df = df[['OriginalTweet']]
df.head()
```

```python
# remove urls
import re

def remove_urls(text):
    try:
        if isinstance(text, str):
            return re.sub(r'http\S+|www\.\S+', '', text)
        return text
    except Exception as e:
        print(f"Error while removing URL: {e}")
        return text
    
df['OriginalTweet'] = df['OriginalTweet'].apply(remove_urls)
```

```python
sentences = [line.split() for line in df['OriginalTweet']]
w2v = Word2Vec(sentences,  window = 5, workers = 4, min_count = 5)
```

```python
words = list(w2v.wv.vocab)
words
```
