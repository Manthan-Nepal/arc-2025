```python
import pandas as pd
import numpy as np
```

```python
# creating an empty dicitionary
dict_data = {
    'name' : [],
    'roll' : [],
    'address' : [],
    'age' : []
}
```

```python
# Ask User for 10 valid inputs

for i in range(10):
    print(f"Enter details for the {i+1}th student :\n")
    name = input("Enter the name\n")
    roll = input("Enter the roll no\n")
    address = input("Enter the address\n")
    
    while True:
        try:
            age = int(input("Enter the age\n"))
            break
        except ValueError:
            print("Please enter the valid age")
    
    dict_data['name'].append(name)
    dict_data['roll'].append(roll)
    dict_data['address'].append(address)
    dict_data['age'].append(age)
    
```

```python
df = pd.dataframe()
df['category'] = df['age'].apply(
    lambda age: 'Junior' if age <= 15 else 'Mid' if age <= 25 else 'Senior'
)
```

```python
print("The dataframe is :")
print(df)

# calculate the average of the age

avg_age = df['age'].mean()
print(f"The average age is {avg_age}\n")

# Calculate the most frequent category

most_fre_category = df['category'].mode()[0]
print(f"The most frequent category is {most_fre_category}")
```
