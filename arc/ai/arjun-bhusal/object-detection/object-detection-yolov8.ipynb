```python
# Import YOLO class
from ultralytics import YOLO

# Load a pre-trained YOLOv8n model
model = YOLO('yolov8n.pt')  # YOLOv8 Nano version (fast, lightweight)
```

```python
# Train on custom dataset
results = model.train(
    data='dataset/yolo-dataset/data.yaml',  # Path to YAML file
    epochs=10,
    imgsz=640,
    batch=32,  # adjust based on your hardware
    project='custom_yolo_project',
    name='yolov8n_detect',
    verbose=True
)
```

```python
# Validate the model performance on the validation set
metrics = model.val()
```
