### 2.4 Given a list of numbers, find the greatest number.

```python
def greatest_num(num):
    greatest = num[0]
    for i in num[1:]:
        if greatest <= i:
            greatest = i
    return(greatest)
num = [3,5,10,6,9]
greatest = greatest_num(num)
print(f"The greatest number is {greatest}")
    
```

### 2.5 Given a list of integers x, find the peak elements. A peak is an element that is greater than its neighbors.

 #### Example: x = [1, 2, 3, 1, 10, 8]
 #### Output: peaks = [3, 10]
 

```python
def calculate_peaks(num):
    peaks = []
    n = len(num)
    if n == 0:
        return peaks
    if n == 1 or num[0] > num[1]:
        peaks.append(num[0])
    for i in range(1, n - 1):
        if num[i] > num[i - 1] and num[i] > num[i + 1]:
            peaks.append(num[i])
    if num[-1] > num[-2]:
        peaks.append(num[-1])
    return peaks

num = [1,2,3,1,10,8,12]
peaks = calculate_peaks(num)
print(peaks)
    
```

### 2.8 Given a roman numeral, convert it to an integer. Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

 Input: s = "LVIII" <br>
 Output: 58<br>
 Explanation: L = 50, V= 5, III = 3<br>

```python
def roman_to_int(s):
    roman_map = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000
    }

    total = 0
    prev_value = 0

    for char in reversed(s):  # Iterate from right to left
        current = roman_map[char]
        if current < prev_value:
            total -= current
        else:
            total += current
        prev_value = current

    return total

s = "LVIII"
print("Roman numeral:", s)
print("Integer value:", roman_to_int(s))
```

### 2.9 Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid. An input string is valid if:

Open brackets must be closed by the same type of brackets.

Open brackets must be closed in the correct order.

Every close bracket has a corresponding open bracket of the same type.

  Example 1:
  Input: s = "( )"
  Output: true

  Example 2:
  Input: s = "( )[ ]{ }"
  Output: true

  Example 3:
  Input: s = "( ]"
  Output: false

```python
def check_validity(s):
    stack = []
    matching = {')': '(', ']': '[', '}': '{'}

    for char in s:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack[-1] != matching[char]:
                return False
            stack.pop()
    return not stack
    
print(check_validity('[[()]]'))
```

### 2.10Write a program to calculate the frequency of each characters in a string.

Example:<br>
Input: s = "hello"<br>
Output: {
        'h': 1,
        'e': 1,
        'l': 2,
        'o': 1
}

```python
def character_frequency(s):
    freq = {}
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

s = "hello"
print(character_frequency(s))
```

### 2.11 Write a program to generate percentage of each students whose records are given below:

grades_dict = {<br>
        "Alice": {"Math": 90, "Science": 85, "Literature": 88},<br>
        "Bob": {"Math": 78, "Science": 82, "Literature": 80},<br>
        "Charlie": {"Math": 92, "Science": 91, "Literature": 94}<br>
}
<br>

Output: <br>
        Alice: 87.67<br>
        Bob: 80.00<br>
        Charlie: 92.33

```python
def calculate_percentage(grades):
    for students in grades:
        sum = 0
        for subjects in grades[students]:
            sum += grades[students][subjects]
        print(f"{students}:{sum/len(grades[students]):.2f}")
        
grades_dict = {
"Alice": {"Math": 90, "Science": 85, "Literature": 88},
"Bob": {"Math": 78, "Science": 82, "Literature": 80},
"Charlie": {"Math": 92, "Science": 91, "Literature": 94}
}
       
calculate_percentage(grades_dict)
```

### 3.1 You are given a large integer represented as an integer list of digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's. Increment the large integer by one and return the resulting ;list of digits. Implement the logic by creating a function.

 Example 1:
 Input: digits = [1,2,3]
 Output: [1,2,4]
 Explanation: The array represents the integer 123.
 Incrementing by one gives 123 + 1 = 124.
 Thus, the result should be [1,2,4].

 Example 2:
 Input: digits = [4,3,2,1]
 Output: [4,3,2,2]
 Explanation: The array represents the integer 4321.
 Incrementing by one gives 4321 + 1 = 4322.
 Thus, the result should be [4,3,2,2].

 Example 3:
 Input: digits = [9]
 Output: [1,0]
 Explanation: The array represents the integer 9.
 Incrementing by one gives 9 + 1 = 10.
 Thus, the result should be [1,0].

```python
def plus_one(digits):
    n = len(digits)
    
    # Start from the last digit and move backward
    for i in range(n - 1, -1, -1):
        if digits[i] < 9:
            # Just increment and return early
            digits[i] += 1
            return digits
        else:
            # Set current digit to 0 and continue
            digits[i] = 0
    
    # If all digits were 9, we end up with all 0s, so we add 1 at the beginning
    return [1] + digits
print(plus_one([4,3,2,1]))
```

### 4.1 Create a class Rectangle. Define the required object attributes/state yourself. Define class methods: area() and perimeter() and use them by creating an instance of rectangle.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


my_rectangle = Rectangle(10, 5)

# Use the area() and perimeter() methods
print("Area of rectangle:", my_rectangle.area())
print("Perimeter of rectangle:", my_rectangle.perimeter())
```
