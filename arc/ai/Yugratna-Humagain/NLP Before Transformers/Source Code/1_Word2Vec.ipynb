# Word2Vec Implementation

Word2Vec is a technique for natural language processing that represents words as vectors. This notebook demonstrates the implementation of Word2Vec using Python.

## Key Concepts:
1. Continuous Bag of Words (CBOW)
2. Skip-gram model
3. Word embeddings

```python
# Import required libraries
import numpy as np
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# Download required NLTK data
nltk.download('punkt')

# Sample text
text = """Natural language processing (NLP) is a field of artificial intelligence 
that gives computers the ability to understand text and spoken words in much 
the same way human beings can. NLP combines computational linguistics with 
statistical models, machine learning and deep learning."""

# Tokenize the text
sentences = sent_tokenize(text)
word_lists = [word_tokenize(sentence.lower()) for sentence in sentences]
```

```python
print("Tokenized sentences:")
for sent in word_lists:
    print(sent)

# Train Word2Vec model
model = Word2Vec(sentences=word_lists, 
                 vector_size=100,  # Dimensionality of word vectors
                 window=5,         # Maximum distance between current and predicted word
                 min_count=1,      # Ignore words that appear less than this
                 workers=4)        # Number of CPU threads to use during training
```

```python
# Get vocabulary size
vocab_size = len(model.wv.key_to_index)
print(f"Vocabulary size: {vocab_size}")

# Print some vocabulary words
print("\nSome words in vocabulary:")
print(list(model.wv.key_to_index.keys())[:10])

# Find similar words
try:
    similar_words = model.wv.most_similar('language')
    print("Words most similar to 'language':")
    for word, score in similar_words:
        print(f"{word}: {score:.4f}")
except KeyError:
    print("Word not in vocabulary")
```

```python
# Get word vector
word = 'language'
if word in model.wv:
    vector = model.wv[word]
    print(f"Vector for '{word}':\n")
    print(f"Shape: {vector.shape}")
    print("\nFirst 10 dimensions:")
    print(vector[:10])
```

```python
# Larger text corpus
corpus = """
Machine learning is a subset of artificial intelligence that focuses on training computers to learn from data.
Deep learning is a type of machine learning that uses neural networks with multiple layers.
Natural language processing helps computers understand and generate human language.
Computer vision is the field of AI that enables computers to understand visual information from the world.
Data science combines statistics, programming, and domain expertise to extract insights from data.
Neural networks are computing systems inspired by biological neural networks in animal brains.
Artificial intelligence aims to create machines that can simulate human intelligence.
"""

# Tokenize the corpus
sentences = sent_tokenize(corpus)
word_lists = [word_tokenize(sentence.lower()) for sentence in sentences]

# Train a new model
new_model = Word2Vec(sentences=word_lists,
                     vector_size=100,
                     window=5,
                     min_count=1,
                     workers=4)

# Finding similar words with the new model
words_to_try = ['learning', 'neural', 'data', 'intelligence']

for word in words_to_try:
    try:
        similar = new_model.wv.most_similar(word)
        print(f"\nWords most similar to '{word}':")
        for w, score in similar:
            print(f"{w}: {score:.4f}")
    except KeyError:
        print(f"'{word}' not in vocabulary")
```
