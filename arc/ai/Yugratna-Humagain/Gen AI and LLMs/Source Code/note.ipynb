# Discriminative vs. Generative Models


| **Feature**              | **Discriminative Models**                                  | **Generative Models**                                             |
|--------------------------|-------------------------------------------------------------|-------------------------------------------------------------------|
| **Purpose**              | Classify input into predefined categories                   | Generate new data similar to training data                        |
| **Learning Objective**   | Learns boundary between classes ($P(y \mid x)$)             | Learns full data distribution ($P(x, y)$)                         |
| **Example Task**         | Classifying images (e.g., cat vs. dog)                      | Generating new images, text, audio, or code                       |
| **Techniques**           | Logistic Regression, SVM, Random Forest, Neural Networks    | GANs, VAEs, Diffusion Models, LLMs (e.g., GPT)                    |
| **Data Requirement**     | Requires labeled data                                       | Requires large and diverse training data                          |
| **Strengths**            | High accuracy in classification, interpretable outputs      | Creative and versatile; can produce novel outputs                 |
| **Weaknesses**           | Cannot generate new data                                    | May hallucinate or generate inaccurate outputs                    |
| **Use Cases**            | Fraud detection, risk scoring, sentiment analysis           | Content generation, translation, synthetic data creation          |
| **Output Type**          | Class labels or probability scores                          | Text, images, audio, video, or structured data                    |
| **Training Complexity**  | Typically simpler and faster                                | More complex and resource-intensive                               |
| **Example Algorithms**   | Decision Trees, Logistic Regression                         | GPT, StyleGAN, DALL·E, VQ-VAE                                     |

## Summary

- **Discriminative Models**: Best for classification and prediction tasks.
- **Generative Models**: Ideal for creating new, realistic content.
- **GANs**: Combine both types — generator (creates), discriminator (evaluates).

> Use discriminative models when you want to **analyze or classify**, and generative models when you want to **create or simulate**.


# How Large Language Models (LLMs) Work



---

## AI Hierarchy

| Level               | Description |
|---------------------|-------------|
| **Artificial Intelligence (AI)** | Machines that mimic human intelligence |
| **Machine Learning (ML)** | Learning from data without explicit programming |
| **Deep Learning (DL)** | Neural network-based learning from large data |
| **Large Language Models (LLMs)** | Deep learning models specialized in understanding and generating text |

---

## Key Concepts

### Language Modeling
LLMs predict the next word or token in a sentence based on previous context. This forms the basis for tasks like text completion, translation, and summarization.


### Massive Training Data
LLMs are trained on enormous datasets (web pages, books, code, etc.) to learn diverse language patterns, facts, reasoning, and even basic world knowledge.


### Natural Language Generation (NLG)
LLMs don't just understand text—they generate it. They can write essays, emails, poems, code, and more by continuing from user prompts.


### What Does **GPT** Mean?
- **G**enerative – creates new text
- **P**retrained – trained on vast general data first
- **T**ransformer – uses transformer architecture (attention mechanism)
---

## Phases of Training LLMs

1. **Pretraining**:  
   - Self-supervised learning to predict the next word  
   - Learns general language understanding

2. **Fine-tuning**:  
   - Supervised learning on task-specific or safer data  
   - Often includes **RLHF** (Reinforcement Learning from Human Feedback)

---


## Core Capabilities

| Capability              | Explanation |
|--------------------------|-------------|
| **Zero-shot Prompting**  | Perform a task without examples (just clear instruction). |
| **Few-shot Learning**    | Perform tasks by seeing a few examples in the prompt. |
| **Chain of Thought**     | Models generate step-by-step reasoning for more complex tasks (e.g., math, logic). |

---

> 💡 LLMs don’t *think*—they predict what comes next. But with enough data and training, that prediction mimics intelligence.


# Hallucinations in Large Language Models (LLMs)

---

## What Are “Hallucinations” in LLMs?

A hallucination occurs when an LLM generates content that is factually incorrect, unverifiable, or nonsensical, even though it may appear coherent and confident.

Examples include:
- Fake citations or sources
- Incorrect facts or flawed logic

---

## Why Do LLMs Hallucinate?

| Cause                        | Explanation |
|-----------------------------|-------------|
| Prediction-based Nature     | LLMs are trained to predict the next word, not verify truth. |
| Lack of Real Understanding  | LLMs model patterns in text but don’t possess true comprehension. |
| Training Data Noise         | Inaccurate or biased data introduces errors during generation. |
| Out-of-Distribution Prompts | Prompts beyond the model's training lead to guessing or fabrication. |
| Overconfidence Bias         | LLMs may present incorrect outputs with high confidence. |

---

## Hallucinations in Different Applications

| Application Domain | Example Risks |
|--------------------|---------------|
| Search Engines     | Incorrect answers, fabricated sources |
| Healthcare         | False medical advice or nonexistent treatments |
| Education          | Misleading explanations or fake historical facts |
| Programming        | Nonexistent libraries or syntactically invalid code |

---

## Strategies to Mitigate Hallucinations

| Strategy                          | Description |
|-----------------------------------|-------------|
| Retrieval-Augmented Generation (RAG) | Combine LLMs with trusted external sources during inference |
| Human-in-the-Loop (HITL)          | Involve human reviewers to validate or correct outputs |
| Fine-tuning on Reliable Data      | Train or adapt the model on high-quality, domain-specific data |
| Prompt Engineering                | Craft precise prompts to reduce ambiguity and error |
| Post-processing and Validation    | Use rules or external tools to flag incorrect content |
| Verification Models               | Use separate models to fact-check LLM responses before showing results |

# Prompting Techniques for Generative AI

Prompting is the method of instructing a generative AI model by writing effective input (prompts). The goal is to guide the model to produce the desired output.

---

## 1. What is Prompt Design?

Prompt design is the process of creating clear and effective prompts to get useful responses from generative models. A well-designed prompt increases accuracy, relevance, and control over the output.

---

## 2. Basic Principles

- **Clarity**: Use simple, unambiguous language.
- **Specificity**: Be detailed about the task or format required.
- **Context**: Provide background or examples when needed.
- **Structure**: Organize the prompt clearly (e.g., input → instruction → output format).

---

## 3. Prompting Strategies

| Strategy               | Description |
|------------------------|-------------|
| **Zero-shot prompting** | Ask the model to complete a task with no examples. <br>_Example_: "Translate 'Good morning' to Nepali." |
| **Few-shot prompting**  | Provide a few examples to help the model understand the pattern. <br>_Example_: "Translate these sentences..." followed by examples. |
| **Instruction prompting** | Give direct instructions on what the model should do. <br>_Example_: "Summarize this article in one paragraph." |
| **Chain-of-thought prompting** | Ask the model to explain its reasoning step by step. <br>_Example_: "Explain how you solved this math problem." |
| **Output formatting**   | Specify the desired structure or format of the response. <br>_Example_: "Answer in JSON format." |
| **Role prompting**      | Assign the model a role or persona. <br>_Example_: "You are a helpful customer support agent." |
| **Multi-turn prompting** | Use a conversational sequence to build context over multiple turns. Useful in chat-based models. |

---

## 4. Tips for Better Prompts

- Start simple, then iterate.
- Test different variations.
- Use delimiters (like triple quotes) for clarity.
- Include constraints if needed (e.g., word count, tone).
- Avoid vague questions or open-ended prompts when you want precision.
