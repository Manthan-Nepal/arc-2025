```python
import pandas as pd
```

```python
#Empty dictionary
data = {
    'name': [],
    'roll': [],
    'address': [],
    'age': []
}
```

```python
# Collect data from user
for i in range(10):
    print(f"Enter data for student {i+1}")
    name = input('Enter name: ')
    roll = input('Enter roll number: ')
    address = input('Enter address: ')
    age = int(input('Enter age: '))
    
    print(f'Enter name for student: {name}')
    print(f'Enter roll for student: {roll}')
    print(f'Enter address for student: {address}')
    print(f'Enter age for student: {age}')
    print('---'*15)
    data['name'].append(name)
    data['roll'].append(roll)
    data['address'].append(address)
    data['age'].append(age)  
```

```python
# Convert the data to dataframe
df = pd.DataFrame(data)
```

```python
# Add category column on basis of age 
def categorize(age):
    if age <= 15:
        return 'Junior'
    elif age <= 25:
        return 'Mid'
    else:
        return 'Senior'

df['category'] = df['age'].apply(categorize)
```

```python
print(df)
```

```python
#Average age
average_age = df['age'].mean()
print(f"\nAverage Age: {average_age:.2f}")
```

```python
# Most frequent category
most_frequent_category = df['category'].mode()[0]
print(f"Most Frequent Category: {most_frequent_category}")
```
