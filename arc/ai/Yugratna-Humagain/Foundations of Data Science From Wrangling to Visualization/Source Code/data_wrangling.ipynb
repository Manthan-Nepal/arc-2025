### 2. Data Structuring
- Reshape data (if needed)  
- Convert data types  
- Set correct formats for features  

```python
cat_cols = ['sex', 'smoker', 'region']
num_cols = ['age', 'bmi', 'children', 'charges']
```

```python
#Before
df.dtypes
```

```python
for col in cat_cols:
    df[col] = df[col].astype('category')
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce') 
```

```python
#After
df.dtypes
```

### 3. Data Cleaning
- Identify & handle missing values  
  - No missing values as seen above 
- Check for & remove duplicates  
- Remove irrelevant features  

```python
# Missing values
df.isnull().sum()
```

```python
# Duplicates
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

### 4. Feature Extraction
- Potential feature extraction  
  - Here: create `age_group` & `bmi_category`  

```python
# Age groups
bins_age = [0, 18, 30, 45, 60, np.inf]
labels_age = ['<18','18-30','31-45','46-60','60+']
df['age_group'] = pd.cut(df['age'], bins=bins_age, labels=labels_age)
```

```python
# BMI categories
bins_bmi = [0, 18.5, 24.9, 29.9, np.inf]
labels_bmi = ['Underweight','Normal','Overweight','Obese']
df['bmi_category'] = pd.cut(df['bmi'], bins=bins_bmi, labels=labels_bmi)
```

### 7. Handling Categorical Variables
- Identify categorical columns  
- Check unique values  
- Decide on encoding strategy  

```python
cat_all = ['sex','smoker','region','age_group','bmi_category']
print("Categorical columns:", cat_all)
for col in cat_all:
    print(f"{col} -> unique values: {df[col].unique()}")
```

### 8. Encoding Categorical Variables
- Label Encoding (Applied to binary variables, where a single 0/1 mapping is sufficient and efficient.)
- One-Hot Encoding with `drop_first=True` (Used for categorical variables with multiple classes to prevent the model from assuming any order or hierarchy.)

`drop_first=True` is used to avoid repeating the same information,which helps models make better and more stable predictions.  Only `sex_male` is included and `sex_female` is excluded as including it would introduce redundancy

```python
# !pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encode 'smoker'
le = LabelEncoder()
df['smoker_LE'] = le.fit_transform(df['smoker'])
df[['smoker_LE','smoker']].head()
```

```python
# One-Hot Encode all other categoricals
df_encoded = pd.get_dummies(df, columns=[c for c in cat_all if c!='smoker'], drop_first=True)
df_encoded.head()
```

```python
print(f'Before encoding: {df.shape}')
print(f'After encoding: {df_encoded.shape}')
```
