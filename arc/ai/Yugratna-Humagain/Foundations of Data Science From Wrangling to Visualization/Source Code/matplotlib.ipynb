### Creating Basic Plots
Demonstrating line, scatter, bar, and histogram plots with example data.

```python
x = [1, 2, 3, 4, 5]
y = [2, 1, 3, 5, 4]
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y) # line plot
plt.title('Line Plot Example')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y) # scatter plot
plt.title('Scatter Plot Example')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(x, y) # bar plot
plt.title('Bar Plot Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(y, bins=3) # histogram
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Plotting with NumPy Arrays
Generate sine and cosine curves using NumPy arra

```python
import numpy as np
```

```python
t = np.linspace(0, 2 * np.pi, 100)
sin_t = np.sin(t)
cos_t = np.cos(t)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(t, sin_t, label='sin(t)')
plt.plot(t, cos_t, label='cos(t)')
plt.title('NumPy Array Plot')
plt.xlabel('t')
plt.ylabel('Value')
plt.legend()
plt.show()
```

### Plotting with pandas DataFrames

```python
import pandas as pd

df_sample = pd.DataFrame({
    'A': np.random.randn(50).cumsum(), #.cumsum() is used for cumulative sum or continually adding one thing to the next and so on.
    'B': np.random.randn(50).cumsum(),
    'C': np.random.randn(50).cumsum()
}, index=pd.date_range('2025-01-01', periods=50))
```

```python
df_sample.head()
```

```python
plt.figure(figsize=(10, 5))

ax = df_sample.plot(kind="line", legend=True)
ax.set_title('Time Series Data Plot')

ax.set_xlabel('Date')
ax.set_ylabel('Cumulative Sum')

plt.show()
```
