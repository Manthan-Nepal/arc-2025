## Introduction to Time Series 

Time series forecasting involves analyzing data points collected or recorded sequentially over time. It uses statistical models and machine learning techniques to analyze past observations and make predictions about future patterns and trends. The key aspect of time series data is that observations are time-dependent, meaning each data point can be influenced by previous observations.

### Characteristics of Time Series Data:

1. **Temporal Ordering**
   - Data points are ordered chronologically
   - Order of observations matters
   - Cannot randomly shuffle the data

2. **Time Dependency**
   - Current values depend on past values
   - Exhibits autocorrelation
   - May show patterns over time

3. **Components**
   - Trend: Long-term progression (for eg. Trend of EV)
   - Seasonality: Repeated patterns (for eg. Demand of clothes during dashain)
   - Cyclical: Non-fixed period fluctuations (for eg. real estate market cycle)
   - Random: Unexplained variations (for eg. COVID-19)

## Stationarity in Time Series Analysis

### What is Stationarity?
Stationarity refers to the condition where the statistical properties of a time series remain consistent over time:
- **Mean**: Remains constant
- **Variance**: Stays the same throughout the series
- **Autocorrelation**: Pattern of dependency doesn't change

### Significance of Stationarity

1. **Simplified Analysis**
   - Makes statistical methods more reliable
   - Enables use of standard time series models
   - Reduces complexity in predictions

2. **Reliable Forecasting**
   - Past patterns more likely to repeat
   - Better prediction accuracy
   - More stable confidence intervals

3. **Statistical Assumptions**
   - Meets requirements for many models
   - Validates hypothesis testing
   - Enables proper model selection

4. **Trend and Seasonality Analysis**
   - Easier decomposition of components
   - Better identification of patterns
   - Clearer seasonal adjustments

### Testing for Stationarity
1. **Visual Tests**
   - Plotting the time series
   - Looking at rolling statistics
   - Checking for obvious trends

2. **Statistical Tests**
   - Augmented Dickey-Fuller (ADF) test
   - Phillips-Perron test

### Common solution
- In case of mean:
    - Trend: Y (t) - Y (t-1) 
     For eg monthly sales -> monthly differences in sales
    - Season: Y (t) - Y(t-S)

- In case of variance
    - ARCH/GARCH model 

## Time Series Models Overview

### 1. Moving Average (MA) Model
A model that uses past errors (residuals) to predict future values by taking weighted averages of past forecast errors.

**Assumptions**:
- Errors are independent
- Constant variance in errors
- No autocorrelation between errors

**Parameters**:
- Order (q): Number of lagged error terms

**Strengths**:
- Good for short-term forecasting
- Handles random fluctuations well
- Simple to implement and interpret

### 2. Autoregressive (AR) Model
A model that predicts future values based on a linear combination of past values.

**Assumptions**: 
- Series is stationary
- Past values have a direct influence on future values
- Errors are normally distributed

**Parameters**: 
- Order (p): Number of lagged values to include

**Strengths**: 
- Good at capturing trends
- Works well with data showing strong autocorrelation
- Simple to interpret

### 3. ARMA (Autoregressive Moving Average) Model
Combines both AR and MA models to leverage the strengths of both approaches.

**Assumptions**:
- Stationarity
- Linear relationships between variables
- Constant variance in errors

**Parameters**:
- p: AR order
- q: MA order

**Strengths**:
- More flexible than pure AR or MA
- Better forecasting accuracy
- Handles complex patterns

### 4. ARIMA (Autoregressive Integrated Moving Average) Model
Extension of ARMA that handles non-stationary data through differencing.

**Assumptions**:
- Data becomes stationary after differencing
- No missing values
- Sufficient historical data

**Parameters**:
- p: AR order
- d: Differencing order
- q: MA order

**Strengths**:
- Handles non-stationary data
- Versatile for many types of time series
- Well-established theoretical foundation

### 5. SARIMA (Seasonal ARIMA) Model
ARIMA model with additional seasonal components.

**Assumptions**:
- Clear seasonal pattern exists
- Seasonality is consistent
- Underlying trend can be made stationary

**Parameters**:
- Regular ARIMA parameters (p,d,q)
- Seasonal parameters (P,D,Q,s)

**Strengths**:
- Captures seasonal patterns
- Handles both trend and seasonality
- Good for regular periodic data

### 6. Exponential Smoothing Models
Family of models that weight recent observations more heavily than older ones.

**Assumptions**:
- More recent observations are more relevant
- Components (trend, seasonal) are additive or multiplicative

**Parameters**:
- Alpha: Level smoothing
- Beta: Trend smoothing
- Gamma: Seasonal smoothing

**Strengths**:
- Simple to understand and implement
- Works well with clear trends
- Good for short-term forecasting

### 7. Vector Autoregression (VAR) Model
Multivariate extension of AR models that captures relationships between multiple time series.

**Assumptions**:
- All variables influence each other
- Series are stationary
- Linear relationships

**Parameters**:
- Order (p): Lag order
- Number of variables

**Strengths**:
- Captures interdependencies
- Good for system analysis
- Handles multiple variables

### 8. Machine Learning Models
Modern approaches using algorithms like Neural Networks, Random Forests, and XGBoost.

**Assumptions**:
- Sufficient training data
- Pattern consistency
- Computational resources available

**Parameters**:
- Model-specific hyperparameters
- Feature engineering choices
- Training parameters

**Strengths**:
- Can capture non-linear relationships
- Flexible and adaptable
- Handle large amounts of data

### Metrics for evaluating time series models

1. Mean Absolute Error (MAE)
   - Sum of absolute differences
   - Same unit as original data
   - Less affected by outliers
   - Easy to interpret

2. Root Mean Squared Error (RMSE)
   - Square root of MSE
   - Penalizes large errors
   - Units match original data
   - Standard in forecasting

3. Mean Absolute Percentage Error (MAPE)
   - Percentage-based measure
   - Independent of scale
   - Easy to compare different series
   - Not suitable for zero values

4. Forecast Bias
   - Measures systematic errors
   - Shows over/under prediction
   - Important for business decisions
   - Helps in model adjustment

## Autocorrelation and Partial Autocorrelation

### Autocorrelation (ACF)
Measures the correlation between observations at different time lags in a time series. (direct and indirect effect both)

1. **Definition**
   - Correlation between current and past values
   - Shows dependency over time
   - Range: -1 to +1
   - Lag-k correlation: correlation between Yt and Yt-k

2. **Interpretation**
   - Positive ACF: Similar patterns repeat
   - Negative ACF: Opposite patterns repeat
   - Zero ACF: No relationship
   - Significant spikes indicate seasonality

3. **Uses**
   - Identify seasonality
   - Detect non-randomness
   - Determine model orders
   - Check stationarity

### Partial Autocorrelation (PACF)
Measures the direct correlation between observations at different time lags after removing the effects of intermediate observations.

1. **Definition**
   - Direct correlation at each lag
   - Removes intermediate effects
   - More specific than ACF
   - Helps identify AR order

2. **Interpretation**
   - Positive PACF: Direct positive relationship at that lag
   - Negative PACF: Direct negative relationship at that lag
   - Zero PACF: No direct relationship
   - Sharp cut-off indicates potential AR order
