# Foundations of Natural Language Processing (NLP)

### 1. What is NLP?
Natural Language Processing (NLP) is the technology that allows computers to read, understand, and generate human language. It bridges human communication and computer understanding.

### 2. Benefits of NLP
- **Automation:** Quickly process large amounts of text.  
- **Insights:** Extract sentiment, trends, and key information.  
- **Interaction:** Power chatbots and virtual assistants.  
- **Accessibility:** Convert speech to text and vice versa.

### 3. Approaches to NLP
- **Rule-Based:** Hand-crafted language rules.  
- **Statistical:** Models learn from large text samples (e.g., n-gram).  
- **Machine Learning:** Algorithms trained on labeled data (e.g., classifiers).  
- **Deep Learning:** Neural networks capture complex patterns (e.g., transformers).

### 4. NLP Tasks
- **Text Classification:** e.g., spam detection.  
- **Named Entity Recognition (NER):** Identify names, places.  
- **Sentiment Analysis:** Determine positive/negative tone.  
- **Machine Translation:** Translate text between languages.  
- **Question Answering:** Find answers in documents.

### 5. How NLP Works
1. **Preprocessing:** Clean and prepare text.  
2. **Representation:** Convert words into numbers.  
3. **Modeling:** Use algorithms to learn patterns.  
4. **Postprocessing:** Convert model output back to text or labels.

### 6. Challenges of NLP
- **Ambiguity:** Words can have multiple meanings.  
- **Context:** Understanding depends on surrounding words.  
- **Data Quality:** Noisy, unstructured text.  
- **Language Diversity:** Dialects, slang, and jargon.

### 7. NLP Use Cases
- **Healthcare:** Extract patient info from records.  
- **Finance:** Analyze market news sentiment.  
- **Retail:** Automate customer support via chatbots.  
- **Marketing:** Monitor brand sentiment on social media.

## Key Terms & Concepts
- **Corpus**: A large and structured set of texts used for training and evaluating NLP models.
- **Token**: A single unit (word, subword, character) extracted from text during tokenization.
- **Vocabulary**: The set of all unique tokens in a corpus.
- **N‑gram**: Contiguous sequence of *n* tokens (e.g., bigram, trigram).
- **Part‑of‑Speech (POS) Tagging**: Assigning word classes (noun, verb, adjective) to tokens.
- **Syntax vs. Semantics**: Syntax relates to structure; semantics relates to meaning.

# Preprocessing Steps for NLP

1. **Text Cleaning**  
   - Convert to lowercase  
   - Remove URLs  
   - Remove non-word and non-whitespace characters  
   - Remove digits

2. **Tokenization**  
   - Split text into words or subwords (tokens).

3. **Stopword Removal**  
   - Remove common words (e.g., "the", "is").

4. **Stemming/Lemmatization**  
   - **Stemming:** Cut words to their root form (e.g., "running" → "run").  
   - **Lemmatization:** Convert words to base dictionary forms (e.g., "better" → "good").


# Tokenizers in NLP

### What Is Tokenization?
Tokenization is the process of breaking down text into smaller units, often called tokens. These tokens can be words, characters, or subwords, depending on the type of tokenizer used. This step is essential because machine learning models require numerical data as input, and tokenizers bridge the gap between human-readable text and model-friendly numbers.

### Why Is Tokenization Important?
- Structure raw text
- Preserve meaning 
- Standardize input

### Types of Tokenization
- **Word-Based:** Split on spaces and punctuation.  
- **Character-Based:** Each character is a token.  
- **Subword:** Use units between words and characters (e.g., BPE, WordPiece).

# Types of Text Representation



## One-Hot Encoding
- Represent each token as a binary vector with one ‘1’ at its token index.

## Bag of Words (BoW)
- Count how many times each token appears in a document.

## TF-IDF
- **TF:** Term Frequency—the count of a term in a document.  
- **IDF:** Inverse Document Frequency—downweight common terms.  
- **TF-IDF:** TF × IDF gives importance to rare but frequent-in-doc terms.

## Word2Vec
- Neural embeddings that place similar words close together in vector space.  
- Two models: CBOW (predict word from context) and Skip-Gram (predict context from word).

## GloVe
- Builds embeddings by factorizing a global word–word co-occurrence matrix.
