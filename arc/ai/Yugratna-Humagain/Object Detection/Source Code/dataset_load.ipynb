### Image Annotation Formats

| Format      | File Type | Bounding Box Format      | Normalized | Structure | Segmentation Support | Pros                                 | Cons                             |
|-------------|-----------|--------------------------|------------|-----------|-----------------------|--------------------------------------|----------------------------------|
| YOLO        | `.txt`    | `x_center y_center w h`  | Yes          | Simple    | No                     | Lightweight, easy to parse           | No labels/metadata included      |
| Pascal VOC  | `.xml`    | `xmin ymin xmax ymax`    | No          | Medium    | No                     | Structured, human-readable           | Verbose XML format               |
| COCO        | `.json`   | `x y width height`       | No          | Complex   | Yes                     | Supports segmentation, rich metadata | Complex to parse manually        |
