### 1. What is Non-Max Suppression (NMS) and how does it work?

**Non-Max Suppression (NMS)** is a post-processing technique used in object detection to eliminate redundant bounding boxes that refer to the same object.

#### How it works:
1. **Start with predicted bounding boxes**: Each with a confidence score and class label.
2. **Sort** the boxes by their confidence scores in descending order.
3. **Pick the box** with the highest confidence score.
4. **Suppress** all other boxes that have a high Intersection over Union (IoU) overlap (usually > 0.5) with the selected box.
5. **Repeat** the process for the remaining boxes.

This helps retain only the most relevant bounding box for each detected object, reducing duplicates and improving precision.

#### 1. Input: Bounding Boxes and Scores

We have 5 bounding boxes, each with an associated confidence score:

| Box ID | x1 | y1 | x2 | y2 | Score |
|--------|----|----|----|----|-------|
| B1     | 10 | 10 | 50 | 50 | 0.9   |
| B2     | 20 | 20 | 60 | 60 | 0.8   |
| B3     | 15 | 15 | 55 | 55 | 0.85  |
| B4     | 40 | 40 | 80 | 80 | 0.7   |
| B5     | 30 | 30 | 70 | 70 | 0.6   |

#### 2. Sort Boxes by Confidence Score

Sort the bounding boxes by their score in descending order:

- **B1** (Score = 0.9)
- **B3** (Score = 0.85)
- **B2** (Score = 0.8)
- **B4** (Score = 0.7)
- **B5** (Score = 0.6)

#### 3. Initialize List for Selected Boxes

Create an empty list to store boxes that we will keep.

#### 4. Select the Box with the Highest Score

Select the box with the highest score (`B1 = 0.9`) and add it to the list.

**Selected list**: `[B1]`

#### 5. Compute Intersection over Union (IoU) for Remaining Boxes

Calculate the IoU (overlap) between the selected box (B1) and the rest of the boxes:

- IoU(B1, B2) = 0.6 (Discard B2, overlap too high)
- IoU(B1, B3) = 0.5 (Discard B3, overlap too high)
- IoU(B1, B4) = 0.2 (Keep B4, overlap low)
- IoU(B1, B5) = 0.3 (Keep B5, overlap low)

#### 6. Discard Boxes with IoU Above Threshold

- Set IoU threshold = 0.5.
- Discard B2 and B3 because their IoU with B1 is above 0.5.
- Keep B4 and B5 as their IoU values are below the threshold.

#### 7. Repeat for Remaining Boxes

Now, take the next highest score, which is **B4** (score = 0.7), and add it to the selected list.

Then calculate IoU between B4 and the remaining box B5:

- **IoU(B4, B5) = 0.3** (Keep B5, overlap is low).

**Selected list**: `[B1, B4, B5]`

#### 8. Output the Selected Boxes

After applying NMS, the final selected bounding boxes are:

- **B1** (score = 0.9)
- **B4** (score = 0.7)
- **B5** (score = 0.6)

These are the boxes that we keep.

### 2. What are the evaluation metrics used for object detection tasks? Is there any relationship between those metrics and IoU?

#### Common Evaluation Metrics:

- **Intersection over Union (IoU)**: Measures the overlap between predicted and ground truth bounding boxes.

  $$
  \text{IoU} = \frac{\text{Area of Overlap}}{\text{Area of Union}}
  $$

- **Precision**: Fraction of correct positive predictions among all positive predictions.

  $$
  \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
  $$

- **Recall**: Fraction of correct positive predictions among all actual positives.

  $$
  \text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
  $$

- **Average Precision (AP)**: Area under the Precision-Recall curve for a specific class.

- **Mean Average Precision (mAP)**: Mean of AP across all classes, often evaluated at specific IoU thresholds (e.g., mAP@0.5, mAP@[0.5:0.95]).


#### Relationship with IoU:

- IoU is used as a **criterion to determine true positives**.
- A prediction is considered a true positive if its IoU with the ground truth is greater than a specified threshold (typically 0.5).
- Increasing the IoU threshold makes detection stricter, reducing the number of true positives and potentially lowering metrics like Recall and mAP.
