# Responsible AI

## Chapter 1: Understanding Security and Responsible AI

### 1.1 Responsible AI Principles and Frameworks
- **Fairness**  
  - Definition: AI must treat all groups equitably.  
  - Example: Use demographic-balanced training data.
- **Reliability & Safety**  
  - Definition: Models produce accurate, predictable results.  
  - Example: Validate data quality and implement fallback logic.
- **Privacy & Security**  
  - Definition: Protect data confidentiality and integrity.  
  - Example: Encrypt PII at rest and in transit.
- **Inclusiveness**  
  - Definition: Design for diverse abilities and contexts.  
  - Example: Support screen readers and multilingual UX.
- **Transparency**  
  - Definition: Explain how outputs are derived.  
  - Example: Provide model cards or decision logs.
- **Accountability**  
  - Definition: Assign clear human oversight.  
  - Example: Define roles for monitoring and incident response.

### 1.2 Microsoft’s Responsible AI Approach
- **Standard Playbook**: Embeds the six principles across design, build, and deployment phases.
- **Collaborative Lifecycle**: Continuous feedback loops between engineers, customers, and researchers.

### 1.3 Foundations of Security in AI on Azure

- **Key Controls**  
  - Encrypt data at rest and in transit  
  - Secure pipelines, endpoints, and development environments  
  - Implement monitoring, alerting, and incident response plans

  1. Document use-case goals, data requirements, and failure modes

- **Common Attacks:**  
  - Prompt injection  
  - Evasion  
  - Model inversion  
  - Denial of Service  
  - Impersonation
- **Mitigation:** Defensive coding, rate limits, anomaly detection

### 2.4 Best Practices Overview
1. **Data Minimization:** Only collect data you need.  
2. **Robustness:** Test adversarial scenarios early.  
3. **Bias Mitigation:** Use dashboards and bias-detection tools.  
4. **Explainability:** Surface content credentials.  
5. **Incident Response:** Define playbooks and rollback plans.

## Chapter 4: Implementing Security Measures

### 4.1 Jailbreak Detection for LLMs
- **Attack Types:**  
  - Altered system messages  
  - Embedded confusing prompts  
  - Role-play bypasses  
  - Encoded instructions
- **Mitigation:** Use jailbreak-detection filters on both input and output.

### 4.2 Data Security for LLM Pipelines
- **Authentication:** Microsoft Entra ID and managed identities.  
- **Network Isolation:** Private endpoints, VNETs, private DNS.  
- **Encryption:** TLS for transit; CMKs or KMS in Key Vault for at-rest.

### 4.3 Infrastructure Security (Zero Trust)
- **Verify Explicitly:** RBAC, MFA, conditional access.  
- **Least Privilege:** Limit roles to minimal scopes.  
- **Assume Breach:** Harden endpoints; continuous monitoring.

#### 4.3.1 Azure ML Best Practices
- RBAC for workspaces and compute  
- Private managed VNETs and endpoints  
- Key Vault for secrets  
- Secure model endpoints (token auth, no public access)

#### 4.3.2 Azure AI Services Best Practices
- Managed identities and RBAC for service access  
- Follow security baselines  
- Embed private network access patterns

### 4.4 General Security Tips
- Train teams on security fundamentals  
- Keep humans in the loop for high-risk decisions  
- Apply secure-by-design cybersecurity practices

## Responsible AI: Security and Privacy

### Definitions
- **Security**: Protecting AI systems’ integrity, confidentiality, and availability; ensuring safe interactions among user prompts, models, and outputs.
- **Privacy**: Preventing misuse or unintended disclosure of sensitive data during AI training and inference.

### Gen AI Pipeline Phases & Data Protection
1. **Pre-training**  
   - Trains on vast datasets (web pages, code repos) to predict next tokens.  
   - Risk: memorized Personally Identifiable Information (PII) or copyrighted text.  
2. **Post-training / Alignment**  
   - Fine-tuned on instructions, human annotations to follow guidelines and values.  
   - Risk: sensitive data in instruction-response pairs.  
3. **Deployment**  
   - Serves user inputs (text, documents, images, KBs) and returns generated responses.  
   - Risk: leaking system prompts or retrieval-augmented content.

_All data at every phase is potentially disclosable—use encryption, access controls, and sanitization._

### Data Curation & Sanitization
- **Simple Data**: Remove via regex (names, addresses, dates).  
- **Complex Data**: Detect with heuristic/statistical or ML classifiers.  
- **Evaluation**: Test sanitization against manually annotated datasets to minimize false negatives.

### Model Alignment & Fine-Tuning
- **Supervised Fine-Tuning**: Teaches instruction following and refusal of harmful requests.  
- **Reinforcement Learning from Human Feedback (RLHF/DPO)**: Optimizes outputs to align with human preferences and suppress toxic or privacy-violating content.

### Inference-Time Guardrails
- Real-time monitors that detect and block:
  - Sensitive topics or PII disclosures  
  - Toxic or policy-violating content  
- Examples: Amazon Bedrock guardrails, Meta CodeShield  
- Allow customers to customize system prompts and external checks.

### Gen AI Threat Categories
1. **Data Exfiltration**  
   - Theft of training data, prompts, model weights, architectures.
2. **Model Integrity Compromise**  
   - *Data Poisoning*: Inserting malicious data into training or RAG corpora.  
   - *Model Drift*: Performance degradation as real data diverges from training distribution.
3. **Trust & Safety Risks**  
   - Bias, discrimination, misinformation, hallucinations, toxic output.

### Attack Techniques & Mitigations
- **Data Poisoning**  
  - Mitigation: Trusted data curation, adversarial training, continuous validation.
- **Jailbreaking & Prompt Injection**  
  - Techniques: refusal suppression, encoding (Base64, cipher), role-play, token insertion.  
  - Mitigation: Domain-specific guardrails, adversarial testing, aggressive input/output sanitization.
- **Model & Prompt Extraction**  
  - Attack: Reconstruct prompts or infer model parameters from responses.  
  - Mitigation: Avoid sensitive info in prompts, limit API exposure, sanitize training data, use “detect prompt attack” features.
