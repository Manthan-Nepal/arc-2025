## Introduction
With the rise of Artificial Intelligence (AI), UX design is evolving. Text-based and voice-based AI interfaces (Conversational UX) are becoming increasingly common in products like:
- Gmail, LinkedIn – smart replies  
- Slackbot – customer support  
- Google Home, Alexa – voice interaction  
- Facebook – facial recognition  
- Nest – multi-sensory learning  

---

## 1. Define the Personality of the AI Bot

- Give your bot a human-like identity (name, tone, demeanor).
- Ask: If this bot were a person, what would they be like?
- Tailor personality based on the target user:
  - Teaching kids? Use playful, colorful, friendly tone.
  - Helping professionals? Use elegant, patient, informed tone.

**Example:**  
Mitsuku Chatbot – Friendly, cute, girl-next-door style personality.  
Not a scar-faced 200 lb man (less inviting for casual conversations).

---

## 2. Design the Activation Experience

- Define clear activation words (e.g., “Hey Google”, not “Hi Google”).
- Provide feedback upon activation:
  - Light ring animation  
  - Verbal cues like “How can I help?”

Goal: Mimic natural human interaction. Avoid forcing users to repeat wake words.



---

## 3. Design the Starting Point & Clear Options

- Offer clear entry points for interaction.
- Show examples of what the bot can do.

**Example (Loan service chatbot):**
- “Get a quote”  
- “How much you can borrow”  
- “Loan options”  

---

## 4. Identify the Primary Engagement Path

- Understand the context in which users interact with the bot.
- Analyze user needs and scenarios.
- Predict typical user questions.

---

## 5. Scriptwriting – Design the Conversation

**a. Draw the Dialog Flow**  
Plan how the conversation will move step-by-step.



**b. Summarize User Input Grammar**  
Users ask the same thing in different ways:
- “What’s traffic like today?”
- “How long to drive home?”
- “Is 101 slow right now?”

**c. Entity Recognition**  
Identify structured data in queries:
- `<Place>`, `<Time>`, `<Traffic>`  
  Example: “Traffic to <Millbrae> at <Now>”

**d. Contextual Responses**  
- Don’t overload users with data. Summarize and personalize.
- Instead of reading entire news content, offer:
  - Local headlines  
  - Tailored content based on user interests and location

---

## 6. Does the Bot Need Extra Logic?

- Handle ambiguity or misinterpretation:
  - “Turn off” vs. “Turn on”
- Use state awareness to clarify commands:
  - If light is already off and user says “Turn off”, bot might infer they meant “on”.

---

## 7. Role-Playing & Prototyping

- Use low-fi methods like messaging apps for early testing.
- One person plays the bot, the other plays the user.
- Test:
  - Script quality  
  - Coverage of possible questions  
  - Bot tone and behavior

---

## 8. Define Success Metrics

**Quantitative:**
- CPS (Conversations per Session)
- Bot-to-Agent Rate

**Qualitative:**
- Comfort level of users
- Naturalness of conversation

---

## 9. User Privacy vs. Smart AI

- Smart AI often surprises users with what it knows:
  - Example: Google Maps knowing flight details from Gmail
- Transparency is key:
  - Let users know how data is collected and used
  - Respect privacy boundaries to maintain trust
