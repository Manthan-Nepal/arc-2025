## What is a Seq2Seq Model?

Sequence-to-sequence (seq2seq) models are used to transform one sequence into another.  
They are widely used in NLP tasks like:

- Machine Translation
- Text Summarization
- Chatbots

### Components:
- **Encoder:** Converts input sequence into a context vector.
- **Decoder:** Uses context to produce output sequence.

---

## Types of Seq2Seq Models

### 1. One-to-One

- **Description:** Single input ➝ Single output  
- **Example Tasks:** Image Classification, Text Classification  
- **Use Case:** Simple tasks with one input to one output mapping.

---

### 2. One-to-Many

- **Description:** One input ➝ Multiple outputs  
- **Example Task:** Image Captioning  
- **Use Case:** Input expansion like text generation.

---

### 3. Many-to-One

- **Description:** Multiple inputs ➝ Single output  
- **Example Task:** Sentiment Analysis  
- **Use Case:** Sequence summarization to a single output.

---

### 4. Many-to-Many

- **Description:** Multiple inputs ➝ Multiple outputs  
- **Example Task:** Machine Translation (Translating sentences from one language to another)
- **Use Case:** Complex sequence mappings of variable length.

---

## Applications of Seq2Seq Models

- **Machine Translation**
- **Chatbots**
- **Summarization**
- **Speech Recognition**
