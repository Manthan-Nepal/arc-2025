# Unit 1: Introduction to AI Agents

## What Is an Agent?
- An AI agent combines a reasoning component (an LLM) with tools that enable actions.
- The agent operates by repeating a cycle of thinking about user intent, invoking tools, and integrating results.

## What Are LLMs?
- Large Language Models are trained on extensive text corpora to predict and generate language.
- In agents, LLMs function as the reasoning core: they interpret prompts, plan sequences of actions, and decide when to call tools.

## Messages and Special Tokens
- Agents communicate with LLMs using structured message roles (`system`, `user`, `assistant`) to set context and goals.
- Special tokens mark boundaries for tool invocation or signal shifts between reasoning and execution stages.

## What Are Tools?
- Tools are external functions or APIs (for example, search engines, calculators, file operations) that extend the agent’s capabilities beyond text generation.
- Agents select and sequence tools to accomplish complex tasks that require data retrieval, computation, or interaction with external systems.

## Thought-Action-Observation Cycle
1. **Thought**  
   - Internal reasoning step where the agent reviews context and plans the next operation.
2. **Action**  
   - Invocation of one or more tools to perform the planned operation.
3. **Observation**  
   - Reception of tool outputs, which are merged back into the agent’s context for further reasoning.

### Internal Reasoning and the Re-Act Approach
- The Re-Act pattern interleaves chain-of-thought reasoning with discrete actions, allowing the agent to adjust its plan dynamically based on intermediate results.

### Actions and Environment Engagement
- Each action represents a high-level task, potentially composed of multiple tool calls, that moves the agent closer to its goal.

### Observation and Adaptation
- Observations from tool outputs feed back into the agent’s knowledge, enabling error handling, refinement, and adaptive planning.

## Dummy Agent Library 
- A minimal framework demonstrating how prompts, tool definitions, and response parsing implement the agent loop in code.
- Illustrates the separation between reasoning logic and tool integration.

## smolagents Framework 
- A lightweight library that structures agent logic into distinct components: planners (LLM reasoning), tool registries, and execution loops.
- Supports different agent styles, such as code-generating agents and JSON-based tool-calling agents.

# Unit 2: Frameworks for AI Agents

## 2.1 smolagents

### Purpose of smolagents
- Provides clear abstractions for combining LLM reasoning with external tools, without heavy infrastructure.
- Enables modular definition of tools, planners, and prompt templates.

### Code-Executing vs. JSON-Calling Agents
- **Code-Executing Agents:** Generate and run Python snippets at runtime to perform computations.
- **JSON-Calling Agents:** Emit structured JSON that is parsed into tool calls, ensuring strict schemas and error handling.

### Tool Integration
- Tools are registered via class or decorator patterns and can wrap anything from simple utility functions to external services.

### Retrieval Agents
- Integrate a vector store or search index with the LLM to retrieve and process documents relevant to user queries.

### Multi-Agent Coordination
- Multiple agents, each specialized (for example, search, summarization, analysis), communicate and collaborate to solve composite tasks.

### Vision and Browser Extensions
- Vision agents leverage computer-vision models to interpret image inputs.
- Browser agents employ headless browsing to navigate web content and extract information.

## 2.2 LlamaIndex

### LlamaHub and Data Connectors
- A repository of connectors for various data sources (databases, file systems, APIs) that feed into the index.

### Core Components
- **Indexes:** Structures for storing and querying embeddings or text (vector stores, SQL, text stores).
- **Nodes:** Discrete pieces of information (documents, paragraphs, records).
- **Query Interfaces:** Layers combining LLMs with index lookups to retrieve and process relevant information.

### Tool Usage in Pipelines
- Tools can be embedded within query workflows to call external APIs or perform computations as part of retrieval and generation.

### Agentic Workflows
- Declarative pipelines define a sequence of steps: data ingestion, indexing, retrieval, processing, and output generation, often with support for asynchronous execution.

---

## 2.3 LangGraph

### Overview of LangGraph
- A graph-based organization framework where each node represents a reasoning or action step and edges carry data between nodes.

### Fundamental Building Blocks
- **Nodes:** Units of work (e.g., search, transform, summarize).
- **Edges:** Data channels connecting node outputs to subsequent node inputs.

### Document Analysis Template
- A common graph pattern: ingest document → split into chunks → index → query → refine summary.

# Unit 3: Use Case for Agentic RAG

## Agentic Retrieval-Augmented Generation (RAG)
- Combines retrieval from a vector store with LLM generation under the agent’s control.
- The agent decides which documents to retrieve and uses them as context for answer generation.

## RAG Tool for Domain Content
- Domain content is embedded into a vector database.
- Search and summarization functions become callable tools within the agent’s loop.

## Tool Integration and Coherent Actions
- Retrieval, text processing, and metadata lookup tools are organized into unified actions that the agent executes in sequence.

## Gala Planning Agent 
- An example agent that answers event-planning questions by retrieving guest information, menus, schedules, and related insights.
