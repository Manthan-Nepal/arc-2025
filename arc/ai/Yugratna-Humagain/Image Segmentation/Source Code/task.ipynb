# Notes for Segmentation Tasks

## 1. Image Classification vs. Image Segmentation vs. Object Detection

| Task                | Description                                                                 | Output                                   | Use Case Example                       |
|---------------------|-----------------------------------------------------------------------------|------------------------------------------|----------------------------------------|
| **Image Classification** | Classifies the entire image into one of the categories.                   | Single label per image                    | Dog vs. Cat image classification       |
| **Object Detection**     | Detects and classifies multiple objects with bounding boxes.             | Labels + Bounding Boxes for each object   | Detecting cars and pedestrians         |
| **Image Segmentation**   | Classifies each pixel of the image (fine-grained understanding).         | Pixel-wise labels (segmentation mask)     | Medical imaging, self-driving cars     |


## 2. Relation Between Input Image Size and Output Size in Image Segmentation Models

- In **image segmentation**, the output is a **segmentation mask** with the same **spatial dimensions** (height and width) as the input image.
- This allows **pixel-wise** labeling.
- However, due to **convolutions, pooling, and upsampling**, some architectures (like U-Net or DeepLab) may shrink and then expand the image to recover the original size.

**Key Point**: Output size ≈ Input size (especially in models using skip connections and upsampling).


## 3. What Are Segmentation Masks?

- A **segmentation mask** is an image where **each pixel** represents the **class label** of the corresponding pixel in the input image.
- It has the same height and width as the input image.
- Example:
  - Pixel = 0 → Background
  - Pixel = 1 → Cat
  - Pixel = 2 → Dog

**Purpose**: Used to train and evaluate segmentation models.

## 5. DICE Score vs. IOU Score

### Comparison Table

| Aspect             | DICE Score       | IOU Score (Jaccard)    |
|--------------------|------------------|------------------------|
| **Range**          | 0 → 1            | 0 → 1                  |
| **Focus**          | Overlap emphasis | Intersection/Union     |
| **Sensitivity**    | High for small regions | Balanced for all regions |
| **Typical Use**    | Medical imaging | General object segmentation |
| **Associated Loss**| Dice Loss        | IOU Loss               |


### Formulas

#### DICE Score
$$
\mathrm{DICE} = \frac{2\,\lvert A \cap B\rvert}{\lvert A\rvert + \lvert B\rvert}
$$

#### IOU Score
$$
\mathrm{IOU} = \frac{\lvert A \cap B\rvert}{\lvert A \cup B\rvert}
$$


## Loss Functions for Segmentation

| Loss Function         | Description                           |
|-----------------------|---------------------------------------|
| **Cross‑Entropy**     | Standard pixel‑wise classification    |
| **Dice Loss**         | Complements DICE overlap metric       |
| **IOU Loss**          | Complements IOU overlap metric        |
| **Combo Loss**        | Weighted sum of CE and Dice losses    |

#### Loss Function Formulas

- **Cross‑Entropy Loss**  
  $$
  \mathcal{L}_{CE} = -\sum_{c} y_c \,\log p_c
  $$

- **Dice Loss**  
  $$
  \mathcal{L}_{Dice} = 1 - \frac{2\,|A \cap B|}{|A| + |B|}
  $$

- **IOU Loss**  
  $$
  \mathcal{L}_{IOU} = 1 - \frac{|A \cap B|}{|A \cup B|}
  $$

- **Combo Loss** (with weight α)  
  $$
  \mathcal{L}_{Combo} = \alpha\,\mathcal{L}_{CE} + (1-\alpha)\,\mathcal{L}_{Dice}
  $$



## Evaluation Metrics for Segmentation

| Metric             | Description                              |
|--------------------|------------------------------------------|
| **Pixel Accuracy** | Fraction of correctly labeled pixels     |
| **Mean IOU**       | Average IOU across all classes           |
| **Mean DICE**      | Average DICE score across all classes    |
| **Class Accuracy** | Pixel accuracy computed per class        |

#### Evaluation Metric Formulas

- **Pixel Accuracy**  
  $$
  \mathrm{PA} = \frac{\text{Number of correctly classified pixels}}{\text{Total number of pixels}}
  $$

- **Mean IOU**  
  $$
  \mathrm{Mean\ IOU} = \frac{1}{C}\sum_{i=1}^{C} \frac{|P_i \cap G_i|}{|P_i \cup G_i|}
  $$

- **Mean DICE**  
  $$
  \mathrm{Mean\ DICE} = \frac{1}{C}\sum_{i=1}^{C} \frac{2\,|P_i \cap G_i|}{|P_i| + |G_i|}
  $$

- **Class Accuracy** for class _i_  
  $$
  \mathrm{CA}_i = \frac{\text{Correct pixels in class }i}{\text{Total pixels of class }i}
  $$
